## Project: System Capacity & Care Load Analytics for Unaccompanied Children

### Step 1A — Import Libraries, Set Paths, and Load Raw Dataset

This notebook follows the project guideline section: **Data Ingestion & Structuring**.

The goal of this step is to:
- Load the raw daily time-series dataset
- Inspect dataset shape
- Review column names
- Confirm the file is readable
- Avoid any cleaning or transformation at this stage

```python
# ============================================================
# Step 1A — Import Libraries, Set Paths, and Load Raw Dataset
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

import pandas as pd
import numpy as np
from pathlib import Path

# Display settings for better notebook readability
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
pd.set_option("display.max_colwidth", 80)

# ------------------------------------------------------------
# Project Paths
# ------------------------------------------------------------

# Notebook is inside /notebooks, so project root is one level above
PROJECT_ROOT = Path.cwd().parent

DATA_RAW_DIR = PROJECT_ROOT / "data" / "raw"
DATA_CLEANED_DIR = PROJECT_ROOT / "data" / "cleaned"
DATA_PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

RAW_FILE_PATH = DATA_RAW_DIR / "HHS_Unaccompanied_Alien_Children_Program.csv"

# ------------------------------------------------------------
# Load Raw Dataset
# ------------------------------------------------------------

df_hhs = pd.read_csv(RAW_FILE_PATH)

# ------------------------------------------------------------
# Initial Structure Check
# ------------------------------------------------------------

print("Raw HHS dataset loaded successfully.")
print(f"Dataset shape: {df_hhs.shape[0]:,} rows × {df_hhs.shape[1]:,} columns")

print("\nColumn names:")
for idx, col in enumerate(df_hhs.columns, start=1):
    print(f"{idx}. {col}")

print("\nDataset preview:")
display(df_hhs.head())

print("\nDataset tail:")
display(df_hhs.tail())

print("\nInitial data types:")
display(df_hhs.dtypes.to_frame(name="Data Type"))
```

## Step 1B — Raw Data Structure Audit

This step audits the raw dataset structure before applying any cleaning or transformation.

The goal of this step is to check:
- Total rows and columns
- Fully blank rows
- Missing values by column
- Duplicate full rows
- Duplicate reporting dates
- Raw dataset memory usage
- Basic structural readiness for the next step

No data cleaning, date conversion, sorting, or KPI calculation is performed in this step.

```python
# ============================================================
# Step 1B — Raw Data Structure Audit
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 1B — Raw Data Structure Audit")
print("=" * 70)

# ------------------------------------------------------------
# 1. Basic shape audit
# ------------------------------------------------------------

total_rows, total_columns = df_hhs.shape

print(f"Total rows: {total_rows:,}")
print(f"Total columns: {total_columns:,}")

# ------------------------------------------------------------
# 2. Fully blank rows check
# ------------------------------------------------------------

fully_blank_rows = df_hhs.isna().all(axis=1).sum()

print(f"\nFully blank rows: {fully_blank_rows:,}")

# ------------------------------------------------------------
# 3. Missing values by column
# ------------------------------------------------------------

missing_values_summary = pd.DataFrame({
    "Column Name": df_hhs.columns,
    "Missing Values": df_hhs.isna().sum().values,
    "Missing Percentage": (df_hhs.isna().mean().values * 100).round(2)
})

print("\nMissing values by column:")
display(missing_values_summary)

# ------------------------------------------------------------
# 4. Duplicate full rows check
# ------------------------------------------------------------

duplicate_full_rows = df_hhs.duplicated().sum()

print(f"\nDuplicate full rows: {duplicate_full_rows:,}")

# ------------------------------------------------------------
# 5. Duplicate reporting dates check
# ------------------------------------------------------------

duplicate_dates = df_hhs["Date"].duplicated().sum()

print(f"Duplicate reporting dates: {duplicate_dates:,}")

if duplicate_dates > 0:
    print("\nDuplicate date records:")
    display(df_hhs[df_hhs["Date"].duplicated(keep=False)].sort_values("Date"))
else:
    print("No duplicate reporting dates found.")

# ------------------------------------------------------------
# 6. Raw dataset memory usage
# ------------------------------------------------------------

memory_usage_mb = df_hhs.memory_usage(deep=True).sum() / (1024 ** 2)

print(f"\nRaw dataset memory usage: {memory_usage_mb:.4f} MB")

# ------------------------------------------------------------
# 7. Structural readiness summary
# ------------------------------------------------------------

structure_audit_summary = pd.DataFrame({
    "Audit Check": [
        "Total Rows",
        "Total Columns",
        "Fully Blank Rows",
        "Duplicate Full Rows",
        "Duplicate Reporting Dates",
        "Columns With Missing Values"
    ],
    "Result": [
        total_rows,
        total_columns,
        fully_blank_rows,
        duplicate_full_rows,
        duplicate_dates,
        (missing_values_summary["Missing Values"] > 0).sum()
    ]
})

print("\nRaw structure audit summary:")
display(structure_audit_summary)
```

## Step 1C — Date Conversion Readiness and Date Range Audit

This step follows the project guideline requirement: **Convert Date to datetime format** and prepare for chronological ordering.

The goal of this step is to:
- Test whether all Date values can be parsed correctly
- Convert the Date column into datetime format
- Check the dataset start date and end date
- Check whether the dataset is currently sorted chronologically
- Prepare the dataset for chronological ordering in the next step

No KPI calculation, numeric column cleaning, daily index creation, or visualization is performed in this step.

```python
# ============================================================
# Step 1C — Date Conversion Readiness and Date Range Audit
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 1C — Date Conversion Readiness and Date Range Audit")
print("=" * 70)

# ------------------------------------------------------------
# 1. Create a working copy for structured ingestion
# ------------------------------------------------------------

df_hhs_structured = df_hhs.copy()

print("Working copy created successfully.")
print(f"Working dataset shape: {df_hhs_structured.shape[0]:,} rows × {df_hhs_structured.shape[1]:,} columns")

# ------------------------------------------------------------
# 2. Date parsing test
# ------------------------------------------------------------

date_parse_test = pd.to_datetime(df_hhs_structured["Date"], errors="coerce")

date_parse_errors = date_parse_test.isna().sum()

print(f"\nDate parsing errors: {date_parse_errors:,}")

if date_parse_errors > 0:
    print("\nRows with date parsing errors:")
    display(df_hhs_structured[date_parse_test.isna()])
else:
    print("All Date values parsed successfully.")

# ------------------------------------------------------------
# 3. Convert Date column to datetime
# ------------------------------------------------------------

df_hhs_structured["Date"] = pd.to_datetime(df_hhs_structured["Date"], errors="coerce")

print("\nDate column converted to datetime format.")

# ------------------------------------------------------------
# 4. Date range audit
# ------------------------------------------------------------

start_date = df_hhs_structured["Date"].min()
end_date = df_hhs_structured["Date"].max()
total_reporting_dates = df_hhs_structured["Date"].nunique()

print(f"\nStart date: {start_date.strftime('%B %d, %Y')}")
print(f"End date: {end_date.strftime('%B %d, %Y')}")
print(f"Unique reporting dates: {total_reporting_dates:,}")

# ------------------------------------------------------------
# 5. Current date ordering check
# ------------------------------------------------------------

is_ascending = df_hhs_structured["Date"].is_monotonic_increasing
is_descending = df_hhs_structured["Date"].is_monotonic_decreasing

print("\nCurrent date ordering:")
print(f"Chronological ascending order: {is_ascending}")
print(f"Reverse chronological descending order: {is_descending}")

if is_ascending:
    ordering_status = "Already in chronological ascending order"
elif is_descending:
    ordering_status = "Currently in reverse chronological descending order"
else:
    ordering_status = "Not consistently sorted"

print(f"Ordering status: {ordering_status}")

# ------------------------------------------------------------
# 6. Data types after Date conversion
# ------------------------------------------------------------

print("\nData types after Date conversion:")
display(df_hhs_structured.dtypes.to_frame(name="Data Type"))

# ------------------------------------------------------------
# 7. Date audit summary
# ------------------------------------------------------------

date_audit_summary = pd.DataFrame({
    "Audit Check": [
        "Date Parsing Errors",
        "Start Date",
        "End Date",
        "Unique Reporting Dates",
        "Chronological Ascending Order",
        "Reverse Chronological Descending Order",
        "Ordering Status"
    ],
    "Result": [
        date_parse_errors,
        start_date.strftime("%Y-%m-%d"),
        end_date.strftime("%Y-%m-%d"),
        total_reporting_dates,
        is_ascending,
        is_descending,
        ordering_status
    ]
})

print("\nDate audit summary:")
display(date_audit_summary)
```

## Step 1D — Chronological Sorting and Complete Daily Index Check

This step follows the project guideline requirement: **Ensure chronological ordering** and **Create a complete daily index**.

The goal of this step is to:
- Sort the dataset from oldest reporting date to newest reporting date
- Confirm chronological ordering after sorting
- Create a complete daily date range between the first and last reporting date
- Compare expected calendar days with available reporting dates
- Identify missing reporting dates

No KPI calculation, numeric column cleaning, visualization, or file export is performed in this step.

```python
# ============================================================
# Step 1D — Chronological Sorting and Complete Daily Index Check
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 1D — Chronological Sorting and Complete Daily Index Check")
print("=" * 75)

# ------------------------------------------------------------
# 1. Sort dataset chronologically
# ------------------------------------------------------------

df_hhs_structured = (
    df_hhs_structured
    .sort_values("Date")
    .reset_index(drop=True)
)

print("Dataset sorted chronologically from oldest date to newest date.")
print(f"Sorted dataset shape: {df_hhs_structured.shape[0]:,} rows × {df_hhs_structured.shape[1]:,} columns")

# ------------------------------------------------------------
# 2. Confirm chronological ordering
# ------------------------------------------------------------

is_ascending_after_sort = df_hhs_structured["Date"].is_monotonic_increasing
is_descending_after_sort = df_hhs_structured["Date"].is_monotonic_decreasing

print("\nDate ordering after sorting:")
print(f"Chronological ascending order: {is_ascending_after_sort}")
print(f"Reverse chronological descending order: {is_descending_after_sort}")

# ------------------------------------------------------------
# 3. Preview sorted date range
# ------------------------------------------------------------

print("\nFirst 5 rows after chronological sorting:")
display(df_hhs_structured.head())

print("\nLast 5 rows after chronological sorting:")
display(df_hhs_structured.tail())

# ------------------------------------------------------------
# 4. Create complete daily date range
# ------------------------------------------------------------

start_date = df_hhs_structured["Date"].min()
end_date = df_hhs_structured["Date"].max()

complete_daily_index = pd.date_range(
    start=start_date,
    end=end_date,
    freq="D"
)

expected_calendar_days = len(complete_daily_index)
available_reporting_dates = df_hhs_structured["Date"].nunique()

print("\nComplete daily index check:")
print(f"Start date: {start_date.strftime('%B %d, %Y')}")
print(f"End date: {end_date.strftime('%B %d, %Y')}")
print(f"Expected calendar days: {expected_calendar_days:,}")
print(f"Available reporting dates: {available_reporting_dates:,}")

# ------------------------------------------------------------
# 5. Identify missing reporting dates
# ------------------------------------------------------------

available_date_index = pd.DatetimeIndex(df_hhs_structured["Date"].unique())

missing_reporting_dates = complete_daily_index.difference(available_date_index)

missing_reporting_dates_count = len(missing_reporting_dates)

print(f"Missing reporting dates: {missing_reporting_dates_count:,}")

# ------------------------------------------------------------
# 6. Create missing reporting dates dataframe
# ------------------------------------------------------------

missing_reporting_dates_df = pd.DataFrame({
    "missing_reporting_date": missing_reporting_dates
})

if missing_reporting_dates_count > 0:
    print("\nFirst 10 missing reporting dates:")
    display(missing_reporting_dates_df.head(10))

    print("\nLast 10 missing reporting dates:")
    display(missing_reporting_dates_df.tail(10))
else:
    print("\nNo missing reporting dates found.")

# ------------------------------------------------------------
# 7. Complete daily index audit summary
# ------------------------------------------------------------

daily_index_audit_summary = pd.DataFrame({
    "Audit Check": [
        "Start Date",
        "End Date",
        "Expected Calendar Days",
        "Available Reporting Dates",
        "Missing Reporting Dates",
        "Chronological Ascending After Sort",
        "Reverse Chronological Descending After Sort"
    ],
    "Result": [
        start_date.strftime("%Y-%m-%d"),
        end_date.strftime("%Y-%m-%d"),
        expected_calendar_days,
        available_reporting_dates,
        missing_reporting_dates_count,
        is_ascending_after_sort,
        is_descending_after_sort
    ]
})

print("\nComplete daily index audit summary:")
display(daily_index_audit_summary)
```

## Step 1E — Save Structured Ingestion Outputs

This step completes Notebook 01 by saving the structured ingestion outputs.

The goal of this step is to:
- Save the chronologically sorted structured dataset
- Save the missing reporting dates file
- Save the complete daily index audit summary
- Confirm output file locations

No numeric column cleaning, KPI calculation, logical validation, or visualization is performed in this step.

```python
# ============================================================
# Step 1E — Save Structured Ingestion Outputs
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 1E — Save Structured Ingestion Outputs")
print("=" * 70)

# ------------------------------------------------------------
# 1. Ensure output directories exist
# ------------------------------------------------------------

DATA_CLEANED_DIR.mkdir(parents=True, exist_ok=True)
DATA_PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

print("Output directories verified successfully.")
print(f"Cleaned data directory: {DATA_CLEANED_DIR}")
print(f"Processed data directory: {DATA_PROCESSED_DIR}")

# ------------------------------------------------------------
# 2. Define output file paths
# ------------------------------------------------------------

structured_data_output_path = DATA_CLEANED_DIR / "01_structured_uac_data.csv"
missing_dates_output_path = DATA_PROCESSED_DIR / "01_missing_reporting_dates.csv"
daily_index_summary_output_path = DATA_PROCESSED_DIR / "01_daily_index_audit_summary.csv"

# ------------------------------------------------------------
# 3. Save structured dataset
# ------------------------------------------------------------

df_hhs_structured.to_csv(structured_data_output_path, index=False)

print("\nStructured dataset saved successfully.")
print(f"File: {structured_data_output_path}")

# ------------------------------------------------------------
# 4. Save missing reporting dates
# ------------------------------------------------------------

missing_reporting_dates_df.to_csv(missing_dates_output_path, index=False)

print("\nMissing reporting dates file saved successfully.")
print(f"File: {missing_dates_output_path}")

# ------------------------------------------------------------
# 5. Save daily index audit summary
# ------------------------------------------------------------

daily_index_audit_summary.to_csv(daily_index_summary_output_path, index=False)

print("\nDaily index audit summary saved successfully.")
print(f"File: {daily_index_summary_output_path}")

# ------------------------------------------------------------
# 6. Output verification summary
# ------------------------------------------------------------

saved_outputs_summary = pd.DataFrame({
    "Output Name": [
        "Structured UAC Dataset",
        "Missing Reporting Dates",
        "Daily Index Audit Summary"
    ],
    "File Path": [
        str(structured_data_output_path),
        str(missing_dates_output_path),
        str(daily_index_summary_output_path)
    ],
    "Expected Purpose": [
        "Chronologically sorted dataset for Notebook 02",
        "Transparency file for missing reporting dates",
        "Summary of complete daily index audit"
    ]
})

print("\nSaved outputs summary:")
display(saved_outputs_summary)
```
