### Step 3A — Load Validated Dataset and Confirm Metric Input Readiness

```python
# ============================================================
# Step 3A — Load Validated Dataset and Confirm Metric Input Readiness
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

import pandas as pd
import numpy as np
from pathlib import Path

# Display settings for better notebook readability
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
pd.set_option("display.max_colwidth", 100)

print("Step 3A — Load Validated Dataset and Confirm Metric Input Readiness")
print("=" * 85)

# ------------------------------------------------------------
# 1. Project paths
# ------------------------------------------------------------

PROJECT_ROOT = Path.cwd().parent

DATA_CLEANED_DIR = PROJECT_ROOT / "data" / "cleaned"
DATA_PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

VALIDATED_FILE_PATH = DATA_CLEANED_DIR / "02_validated_uac_data.csv"

print("Project paths configured successfully.")
print(f"Project root: {PROJECT_ROOT}")
print(f"Validated input file: {VALIDATED_FILE_PATH}")

# ------------------------------------------------------------
# 2. Check if validated file exists
# ------------------------------------------------------------

validated_file_exists = VALIDATED_FILE_PATH.exists()

print(f"\nValidated input file exists: {validated_file_exists}")

if not validated_file_exists:
    raise FileNotFoundError(
        f"Validated dataset not found at: {VALIDATED_FILE_PATH}. "
        "Please complete Notebook 02 Step 2E before running Notebook 03."
    )

# ------------------------------------------------------------
# 3. Load validated dataset
# ------------------------------------------------------------

df_hhs_metrics = pd.read_csv(VALIDATED_FILE_PATH)

print("\nValidated dataset loaded successfully.")
print(f"Dataset shape: {df_hhs_metrics.shape[0]:,} rows × {df_hhs_metrics.shape[1]:,} columns")

# ------------------------------------------------------------
# 4. Convert Date column back to datetime after CSV reload
# ------------------------------------------------------------

df_hhs_metrics["Date"] = pd.to_datetime(df_hhs_metrics["Date"], errors="coerce")

date_parse_errors = df_hhs_metrics["Date"].isna().sum()

print(f"\nDate parse errors after loading validated CSV: {date_parse_errors:,}")

if date_parse_errors > 0:
    print("\nRows with Date parse errors:")
    display(df_hhs_metrics[df_hhs_metrics["Date"].isna()])
else:
    print("Date column parsed successfully.")

# ------------------------------------------------------------
# 5. Required columns for derived capacity metrics
# ------------------------------------------------------------

required_metric_input_columns = [
    "Date",
    "Children apprehended and placed in CBP custody*",
    "Children in CBP custody",
    "Children transferred out of CBP custody",
    "Children in HHS Care",
    "Children discharged from HHS Care"
]

required_anomaly_flag_columns = [
    "transfer_exceeds_cbp_custody_flag",
    "discharge_exceeds_hhs_care_flag",
    "logical_constraint_anomaly_flag"
]

print("\nRequired metric input columns:")
for idx, col in enumerate(required_metric_input_columns, start=1):
    print(f"{idx}. {col}")

print("\nRequired anomaly flag columns:")
for idx, col in enumerate(required_anomaly_flag_columns, start=1):
    print(f"{idx}. {col}")

# ------------------------------------------------------------
# 6. Column availability check
# ------------------------------------------------------------

missing_metric_columns = [
    col for col in required_metric_input_columns
    if col not in df_hhs_metrics.columns
]

missing_flag_columns = [
    col for col in required_anomaly_flag_columns
    if col not in df_hhs_metrics.columns
]

print(f"\nMissing metric input columns: {len(missing_metric_columns)}")
print(f"Missing anomaly flag columns: {len(missing_flag_columns)}")

if missing_metric_columns:
    print("Missing metric columns:")
    print(missing_metric_columns)

if missing_flag_columns:
    print("Missing anomaly flag columns:")
    print(missing_flag_columns)

# ------------------------------------------------------------
# 7. Data type review
# ------------------------------------------------------------

print("\nCurrent data types:")
display(df_hhs_metrics.dtypes.to_frame(name="Data Type"))

# ------------------------------------------------------------
# 8. Preview validated input dataset
# ------------------------------------------------------------

print("\nValidated dataset preview:")
display(df_hhs_metrics.head())

print("\nValidated dataset tail:")
display(df_hhs_metrics.tail())

# ------------------------------------------------------------
# 9. Metric input readiness summary
# ------------------------------------------------------------

metric_input_readiness_summary = pd.DataFrame({
    "Readiness Check": [
        "Validated File Exists",
        "Rows Loaded",
        "Columns Loaded",
        "Expected Rows",
        "Expected Columns After Validation Flags",
        "Date Parse Errors",
        "Missing Metric Input Columns",
        "Missing Anomaly Flag Columns",
        "Date Column Type",
        "Dataset Chronological Order"
    ],
    "Result": [
        validated_file_exists,
        df_hhs_metrics.shape[0],
        df_hhs_metrics.shape[1],
        720,
        9,
        int(date_parse_errors),
        len(missing_metric_columns),
        len(missing_flag_columns),
        str(df_hhs_metrics["Date"].dtype),
        df_hhs_metrics["Date"].is_monotonic_increasing
    ]
})

print("\nMetric input readiness summary:")
display(metric_input_readiness_summary)
```

## Step 3B — Create Core Healthcare Capacity Metrics

This step creates the core derived healthcare capacity metrics defined in the project guidelines.

The metrics created in this step are:
- Total System Load = Children in CBP custody + Children in HHS Care
- Net Daily Intake = Children transferred out of CBP custody − Children discharged from HHS Care
- Care Load Growth Rate = Day-over-day percentage change in Total System Load

No backlog indicator, rolling average, KPI summary, visualization, or business insight generation is performed in this step.

```python
# ============================================================
# Step 3B — Create Core Healthcare Capacity Metrics
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 3B — Create Core Healthcare Capacity Metrics")
print("=" * 75)

# ------------------------------------------------------------
# 1. Create working copy for metric engineering
# ------------------------------------------------------------

df_hhs_capacity = df_hhs_metrics.copy()

print("Working capacity metrics copy created successfully.")
print(f"Working dataset shape before metrics: {df_hhs_capacity.shape[0]:,} rows × {df_hhs_capacity.shape[1]:,} columns")

# ------------------------------------------------------------
# 2. Define source columns
# ------------------------------------------------------------

cbp_custody_col = "Children in CBP custody"
hhs_care_col = "Children in HHS Care"
transfer_col = "Children transferred out of CBP custody"
discharge_col = "Children discharged from HHS Care"

print("\nSource columns used for core capacity metrics:")
print(f"CBP custody column: {cbp_custody_col}")
print(f"HHS care column: {hhs_care_col}")
print(f"Transfer column: {transfer_col}")
print(f"Discharge column: {discharge_col}")

# ------------------------------------------------------------
# 3. Create Total System Load
# Guideline formula: CBP Custody + HHS Care
# ------------------------------------------------------------

df_hhs_capacity["total_system_load"] = (
    df_hhs_capacity[cbp_custody_col] + df_hhs_capacity[hhs_care_col]
)

# ------------------------------------------------------------
# 4. Create Net Daily Intake
# Guideline formula: Transfers into HHS − Discharges from HHS
# ------------------------------------------------------------

df_hhs_capacity["net_daily_intake"] = (
    df_hhs_capacity[transfer_col] - df_hhs_capacity[discharge_col]
)

# ------------------------------------------------------------
# 5. Create Care Load Growth Rate
# Guideline definition: Day-over-day percentage change
# ------------------------------------------------------------

df_hhs_capacity["care_load_growth_rate_pct"] = (
    df_hhs_capacity["total_system_load"]
    .pct_change()
    .mul(100)
)

# First row has no previous day for comparison, so growth rate is set to 0 for safe interpretation
df_hhs_capacity["care_load_growth_rate_pct"] = (
    df_hhs_capacity["care_load_growth_rate_pct"]
    .fillna(0)
    .round(2)
)

print("\nCore capacity metrics created successfully:")
print("1. total_system_load")
print("2. net_daily_intake")
print("3. care_load_growth_rate_pct")

# ------------------------------------------------------------
# 6. Metric validation checks
# ------------------------------------------------------------

metric_columns = [
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct"
]

metric_validation_summary = []

for col in metric_columns:
    metric_validation_summary.append({
        "Metric Column": col,
        "Missing Values": df_hhs_capacity[col].isna().sum(),
        "Minimum Value": df_hhs_capacity[col].min(),
        "Maximum Value": df_hhs_capacity[col].max(),
        "Data Type": str(df_hhs_capacity[col].dtype)
    })

metric_validation_summary_df = pd.DataFrame(metric_validation_summary)

print("\nCore metric validation summary:")
display(metric_validation_summary_df)

# ------------------------------------------------------------
# 7. Preview core metric calculations
# ------------------------------------------------------------

core_metric_preview_columns = [
    "Date",
    cbp_custody_col,
    hhs_care_col,
    "total_system_load",
    transfer_col,
    discharge_col,
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "logical_constraint_anomaly_flag"
]

print("\nCore metric calculation preview:")
display(df_hhs_capacity[core_metric_preview_columns].head(10))

print("\nCore metric calculation tail:")
display(df_hhs_capacity[core_metric_preview_columns].tail(10))

# ------------------------------------------------------------
# 8. Core metric engineering summary
# ------------------------------------------------------------

core_metric_engineering_summary = pd.DataFrame({
    "Metric Engineering Check": [
        "Rows Before Metrics",
        "Columns Before Metrics",
        "Rows After Metrics",
        "Columns After Metrics",
        "Metrics Created",
        "Total System Load Missing Values",
        "Net Daily Intake Missing Values",
        "Care Load Growth Rate Missing Values",
        "Rows Deleted or Modified"
    ],
    "Result": [
        df_hhs_metrics.shape[0],
        df_hhs_metrics.shape[1],
        df_hhs_capacity.shape[0],
        df_hhs_capacity.shape[1],
        len(metric_columns),
        int(df_hhs_capacity["total_system_load"].isna().sum()),
        int(df_hhs_capacity["net_daily_intake"].isna().sum()),
        int(df_hhs_capacity["care_load_growth_rate_pct"].isna().sum()),
        0
    ]
})

print("\nCore metric engineering summary:")
display(core_metric_engineering_summary)
```

## Step 3C — Backlog Indicator Foundation

This step creates the foundation for the guideline-defined backlog indicator.

The project guideline defines the Backlog Indicator as: **Sustained positive net intake over time**.

The goal of this step is to create:
- Positive net intake flag
- Net intake direction label
- Cumulative net intake pressure
- Consecutive positive net intake days

No rolling averages, KPI summary, visualization, or business insight generation is performed in this step.

```python
# ============================================================
# Step 3C — Backlog Indicator Foundation
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 3C — Backlog Indicator Foundation")
print("=" * 75)

# ------------------------------------------------------------
# 1. Confirm input dataset from Step 3B
# ------------------------------------------------------------

print("Input dataset for backlog foundation:")
print(f"Rows: {df_hhs_capacity.shape[0]:,}")
print(f"Columns before backlog foundation: {df_hhs_capacity.shape[1]:,}")

# ------------------------------------------------------------
# 2. Create positive net intake flag
# Guideline logic: sustained positive net intake over time
# ------------------------------------------------------------

df_hhs_capacity["positive_net_intake_flag"] = df_hhs_capacity["net_daily_intake"] > 0

# ------------------------------------------------------------
# 3. Create net intake direction label
# ------------------------------------------------------------

df_hhs_capacity["net_intake_direction"] = np.select(
    [
        df_hhs_capacity["net_daily_intake"] > 0,
        df_hhs_capacity["net_daily_intake"] < 0,
        df_hhs_capacity["net_daily_intake"] == 0
    ],
    [
        "Pressure: Transfers Exceed Discharges",
        "Relief: Discharges Exceed Transfers",
        "Balanced: Transfers Equal Discharges"
    ],
    default="Unknown"
)

# ------------------------------------------------------------
# 4. Create cumulative net intake pressure
# ------------------------------------------------------------

df_hhs_capacity["cumulative_net_intake_pressure"] = (
    df_hhs_capacity["net_daily_intake"].cumsum()
)

# ------------------------------------------------------------
# 5. Create consecutive positive net intake days
# ------------------------------------------------------------

positive_group = (
    df_hhs_capacity["positive_net_intake_flag"]
    != df_hhs_capacity["positive_net_intake_flag"].shift()
).cumsum()

df_hhs_capacity["consecutive_positive_net_intake_days"] = (
    df_hhs_capacity
    .groupby(positive_group)["positive_net_intake_flag"]
    .cumsum()
)

df_hhs_capacity.loc[
    ~df_hhs_capacity["positive_net_intake_flag"],
    "consecutive_positive_net_intake_days"
] = 0

df_hhs_capacity["consecutive_positive_net_intake_days"] = (
    df_hhs_capacity["consecutive_positive_net_intake_days"].astype("int64")
)

print("\nBacklog foundation columns created successfully:")
print("1. positive_net_intake_flag")
print("2. net_intake_direction")
print("3. cumulative_net_intake_pressure")
print("4. consecutive_positive_net_intake_days")

# ------------------------------------------------------------
# 6. Backlog foundation validation summary
# ------------------------------------------------------------

backlog_foundation_columns = [
    "positive_net_intake_flag",
    "net_intake_direction",
    "cumulative_net_intake_pressure",
    "consecutive_positive_net_intake_days"
]

backlog_foundation_validation_summary = []

for col in backlog_foundation_columns:
    backlog_foundation_validation_summary.append({
        "Column Name": col,
        "Missing Values": df_hhs_capacity[col].isna().sum(),
        "Data Type": str(df_hhs_capacity[col].dtype)
    })

backlog_foundation_validation_summary_df = pd.DataFrame(backlog_foundation_validation_summary)

print("\nBacklog foundation validation summary:")
display(backlog_foundation_validation_summary_df)

# ------------------------------------------------------------
# 7. Direction distribution
# ------------------------------------------------------------

net_intake_direction_distribution = (
    df_hhs_capacity["net_intake_direction"]
    .value_counts()
    .reset_index()
)

net_intake_direction_distribution.columns = [
    "Net Intake Direction",
    "Record Count"
]

print("\nNet intake direction distribution:")
display(net_intake_direction_distribution)

# ------------------------------------------------------------
# 8. Backlog foundation preview
# ------------------------------------------------------------

backlog_preview_columns = [
    "Date",
    "net_daily_intake",
    "positive_net_intake_flag",
    "net_intake_direction",
    "cumulative_net_intake_pressure",
    "consecutive_positive_net_intake_days",
    "logical_constraint_anomaly_flag"
]

print("\nBacklog foundation preview:")
display(df_hhs_capacity[backlog_preview_columns].head(15))

print("\nBacklog foundation tail:")
display(df_hhs_capacity[backlog_preview_columns].tail(15))

# ------------------------------------------------------------
# 9. Backlog foundation engineering summary
# ------------------------------------------------------------

backlog_foundation_engineering_summary = pd.DataFrame({
    "Backlog Foundation Check": [
        "Rows Before Backlog Foundation",
        "Columns Before Backlog Foundation",
        "Rows After Backlog Foundation",
        "Columns After Backlog Foundation",
        "Backlog Foundation Columns Created",
        "Positive Net Intake Days",
        "Maximum Consecutive Positive Net Intake Days",
        "Rows Deleted or Modified"
    ],
    "Result": [
        720,
        12,
        df_hhs_capacity.shape[0],
        df_hhs_capacity.shape[1],
        len(backlog_foundation_columns),
        int(df_hhs_capacity["positive_net_intake_flag"].sum()),
        int(df_hhs_capacity["consecutive_positive_net_intake_days"].max()),
        0
    ]
})

print("\nBacklog foundation engineering summary:")
display(backlog_foundation_engineering_summary)
```

## Step 3D — Final Backlog Indicator Classification

This step creates the final backlog indicator based on the project guideline definition: **Sustained positive net intake over time**.

The indicator uses consecutive positive net intake days:
- 0 days: No Current Backlog Pressure
- 1–2 days: Short-Term Intake Pressure
- 3–6 days: Sustained Backlog Pressure
- 7+ days: Prolonged Backlog Pressure

No rolling averages, variability analysis, strain window detection, visualization, or KPI summary is performed in this step.

```python
# ============================================================
# Step 3D — Final Backlog Indicator Classification
# Project: System Capacity & Care Load Analytics for Unaccompanied Children
# ============================================================

print("Step 3D — Final Backlog Indicator Classification")
print("=" * 80)

# ------------------------------------------------------------
# 1. Confirm input dataset from Step 3C
# ------------------------------------------------------------

print("Input dataset for final backlog indicator:")
print(f"Rows: {df_hhs_capacity.shape[0]:,}")
print(f"Columns before final backlog indicator: {df_hhs_capacity.shape[1]:,}")

# ------------------------------------------------------------
# 2. Create final backlog indicator
# Guideline logic: sustained positive net intake over time
# ------------------------------------------------------------

df_hhs_capacity["backlog_indicator"] = np.select(
    [
        df_hhs_capacity["consecutive_positive_net_intake_days"] == 0,
        df_hhs_capacity["consecutive_positive_net_intake_days"].between(1, 2),
        df_hhs_capacity["consecutive_positive_net_intake_days"].between(3, 6),
        df_hhs_capacity["consecutive_positive_net_intake_days"] >= 7
    ],
    [
        "No Current Backlog Pressure",
        "Short-Term Intake Pressure",
        "Sustained Backlog Pressure",
        "Prolonged Backlog Pressure"
    ],
    default="Unknown"
)

# ------------------------------------------------------------
# 3. Create numeric backlog severity score
# ------------------------------------------------------------

backlog_severity_mapping = {
    "No Current Backlog Pressure": 0,
    "Short-Term Intake Pressure": 1,
    "Sustained Backlog Pressure": 2,
    "Prolonged Backlog Pressure": 3
}

df_hhs_capacity["backlog_severity_score"] = (
    df_hhs_capacity["backlog_indicator"]
    .map(backlog_severity_mapping)
    .astype("int64")
)

print("\nFinal backlog indicator columns created successfully:")
print("1. backlog_indicator")
print("2. backlog_severity_score")

# ------------------------------------------------------------
# 4. Backlog indicator distribution
# ------------------------------------------------------------

backlog_indicator_distribution = (
    df_hhs_capacity["backlog_indicator"]
    .value_counts()
    .reset_index()
)

backlog_indicator_distribution.columns = [
    "Backlog Indicator",
    "Record Count"
]

print("\nBacklog indicator distribution:")
display(backlog_indicator_distribution)

# ------------------------------------------------------------
# 5. Backlog severity distribution
# ------------------------------------------------------------

backlog_severity_distribution = (
    df_hhs_capacity["backlog_severity_score"]
    .value_counts()
    .sort_index()
    .reset_index()
)

backlog_severity_distribution.columns = [
    "Backlog Severity Score",
    "Record Count"
]

print("\nBacklog severity score distribution:")
display(backlog_severity_distribution)

# ------------------------------------------------------------
# 6. Final backlog indicator validation summary
# ------------------------------------------------------------

final_backlog_indicator_validation_summary = pd.DataFrame({
    "Backlog Indicator Check": [
        "Rows Checked",
        "Columns Before Final Indicator",
        "Columns After Final Indicator",
        "Backlog Indicator Missing Values",
        "Backlog Severity Score Missing Values",
        "Unique Backlog Indicator Labels",
        "Maximum Backlog Severity Score",
        "Rows Deleted or Modified"
    ],
    "Result": [
        df_hhs_capacity.shape[0],
        16,
        df_hhs_capacity.shape[1],
        int(df_hhs_capacity["backlog_indicator"].isna().sum()),
        int(df_hhs_capacity["backlog_severity_score"].isna().sum()),
        df_hhs_capacity["backlog_indicator"].nunique(),
        int(df_hhs_capacity["backlog_severity_score"].max()),
        0
    ]
})

print("\nFinal backlog indicator validation summary:")
display(final_backlog_indicator_validation_summary)

# ------------------------------------------------------------
# 7. Final backlog indicator preview
# ------------------------------------------------------------

final_backlog_preview_columns = [
    "Date",
    "net_daily_intake",
    "positive_net_intake_flag",
    "net_intake_direction",
    "consecutive_positive_net_intake_days",
    "backlog_indicator",
    "backlog_severity_score",
    "logical_constraint_anomaly_flag"
]

print("\nFinal backlog indicator preview:")
display(df_hhs_capacity[final_backlog_preview_columns].head(20))

print("\nFinal backlog indicator tail:")
display(df_hhs_capacity[final_backlog_preview_columns].tail(20))
```
