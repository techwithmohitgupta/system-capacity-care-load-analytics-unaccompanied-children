# Capacity, Backlog and Operational Pressure Indicator Summary

---

## 1. Purpose of Capacity and Pressure Indicator Analysis

After the dataset foundation and quality checks were completed, the next major goal was to convert raw operational counts into meaningful capacity and pressure indicators.

The original dataset contained daily operational fields such as:

- Children in CBP custody
- Children transferred out of CBP custody
- Children in HHS care
- Children discharged from HHS care

These fields are useful individually, but they become much more valuable when converted into analytical indicators.

The project created several important derived indicators to answer business questions such as:

- How much responsibility is currently sitting inside the overall care system?
- Is the system receiving more children than it is discharging?
- Is care load growing or reducing over time?
- Are backlog conditions temporary or sustained?
- Which days show operational pressure?
- Which periods require closer monitoring?

---

## 2. Total System Load

The first major capacity indicator created in the project was **Total System Load**.

### Formula

`Total System Load = Children in CBP Custody + Children in HHS Care`

This metric represents the total number of children under system responsibility across both CBP custody and HHS care.

### Business Meaning

Total System Load is important because it shows the complete care burden across the system.

Looking only at CBP custody or only at HHS care would give an incomplete picture. The system-wide responsibility includes both groups.

This metric supports the required KPI:

**Total Children Under Care**

### Key Results

| Metric | Value |
|---|---:|
| Latest Total Children Under Care | 2,502 |
| Average Total System Load | 6,232.77 |
| Peak Total System Load | 11,762 |
| Minimum Total System Load | 2,002 |

### Interpretation

The system reached a peak total load of **11,762 children**, which indicates that at certain points the care system carried a very high operational responsibility.

The latest total load of **2,502 children** is much lower than the peak, suggesting that overall load reduced significantly by the end of the analysis period.

This shows that the system did not remain under the same level of pressure throughout the full timeline. Instead, system responsibility changed over time, which makes trend analysis and pressure classification important.

---

## 3. Net Daily Intake

The second major indicator created was **Net Daily Intake**.

### Formula

`Net Daily Intake = Children Transferred Out of CBP Custody − Children Discharged from HHS Care`

This metric measures the balance between inflow into HHS care and outflow through discharge.

### Business Meaning

Net Daily Intake answers a very important operational question:

**Is the care system receiving more children than it is releasing through discharge?**

If net intake is positive, transfers into HHS are higher than discharges.  
If net intake is negative, discharges are higher than transfers.  
If net intake is near zero, inflow and outflow are more balanced.

This metric supports the required KPI:

**Net Intake Pressure**

### Key Results

| Metric | Value |
|---|---:|
| Average Net Intake Pressure | -44.74 children/day |
| Total Net Intake | -32,212 |
| Positive Net Intake Days | 238 |
| Positive Net Intake Day Rate | 33.06% |

### Interpretation

The average net intake pressure was **-44.74 children per day**, meaning that across the full analysis period, discharges generally exceeded transfers on average.

The total net intake was **-32,212**, which indicates that cumulative discharge activity exceeded transfer inflow across the full period.

However, the system still experienced **238 positive net intake days**, representing **33.06%** of the reporting timeline.

This means that although the long-term net movement showed relief, many individual days still created intake pressure.

This is important because operational strain does not only depend on the total period result. Short-term positive intake windows can still create staffing, shelter, medical screening, and placement pressure.

---

## 4. Care Load Growth Rate

The third important indicator was **Care Load Growth Rate**.

### Formula

`Care Load Growth Rate = Day-over-day percentage change in Total System Load`

This metric shows how quickly the total system load increases or decreases from one reporting day to the next.

### Business Meaning

Care Load Growth Rate helps identify whether the system is stable or changing quickly.

A stable system has low movement in care load.  
An unstable system has sharp increases or decreases.

This metric supports the required KPI:

**Care Load Volatility Index**

### Key Results

| Metric | Value |
|---|---:|
| Care Load Volatility Index | 1.89 |
| System Load Standard Deviation | 2,918.46 |
| Average Absolute Growth Change | 1.24 |
| Maximum Final Pressure Score | 6.07 |

### Interpretation

The Care Load Volatility Index was **1.89**, showing that daily care load movement had measurable variability.

The system load standard deviation was **2,918.46**, which shows that total system responsibility fluctuated substantially across the analysis period.

This matters because even if the average system load looks manageable, sudden changes in load can create operational difficulty.

Capacity planning should not only depend on average care load. It should also account for how quickly the care load changes.

---

## 5. Backlog Severity and Backlog Accumulation

The project also created backlog-related indicators to identify sustained care pressure.

Backlog pressure is important because short-term intake pressure may be manageable, but sustained pressure can create system strain.

### Business Meaning

Backlog indicators help answer:

- Are positive net intake days becoming repeated?
- Is intake pressure temporary or sustained?
- Are children staying in the system longer than expected?
- Are discharges offsetting incoming flow quickly enough?

This supports the required KPI:

**Backlog Accumulation Rate**

### Key Results

| Metric | Value |
|---|---:|
| Any Backlog Pressure Days | 238 |
| Sustained Backlog Days | 96 |
| Critical Backlog Days | 27 |
| Backlog Accumulation Rate | 13.33% |

### Interpretation

The system experienced **238 days** with some form of backlog pressure.

Out of these, **96 days** were classified as sustained backlog days, and **27 days** reached critical backlog conditions.

The final backlog accumulation rate was **13.33%**, meaning that sustained backlog pressure existed across a meaningful share of the reporting period.

This is important because backlog pressure indicates that inflow and outflow balance was not always smooth.

Even though total net intake across the full period was negative, backlog pressure still appeared across multiple days. This shows why both long-term totals and short-term pressure patterns must be analyzed together.

---

## 6. Discharge Offset Ratio

Another important operational KPI created was **Discharge Offset Ratio**.

### Formula

`Discharge Offset Ratio = Total Discharges from HHS Care / Total Transfers into HHS Care`

This metric measures whether discharges are offsetting incoming transfers.

### Business Meaning

The Discharge Offset Ratio helps explain whether the system is relieving care load effectively.

If the ratio is greater than 1, discharges are higher than transfers.  
If the ratio is below 1, transfers are higher than discharges.  
If the ratio is close to 1, inflow and outflow are balanced.

This supports the required KPI:

**Discharge Offset Ratio**

### Key Results

| Metric | Value |
|---|---:|
| Discharge Offset Ratio | 1.35 |
| Discharge Offset Percentage | 135% |
| Discharge Surplus / Gap | 32,212 |
| Total HHS Discharges | 124,853 |

### Interpretation

The Discharge Offset Ratio was **1.35**, which means that discharges from HHS care exceeded transfers into HHS care across the full analysis period.

The discharge offset percentage was **135%**, indicating strong overall discharge offset capacity at the cumulative level.

The discharge surplus was **32,212**, meaning that total discharges exceeded total transfers by 32,212 children.

This is a positive long-term signal because it suggests the system was able to relieve load over time.

However, this does not remove the need to monitor pressure. Even with strong cumulative discharge offset, short-term pressure days and backlog windows still occurred.

## 10. Capacity and Pressure Relationship

The analysis showed that system pressure is not explained by only one metric.

Operational pressure was influenced by multiple connected signals:

- Total system load
- Net daily intake
- Care load growth rate
- Backlog severity
- Rolling pressure movement
- Volatility
- Sudden spikes
- Prolonged strain windows
- Final pressure classification

This multi-signal approach is stronger than relying on a single KPI.

For example:

- A high system load day may not always be critical if discharges are strong.
- A low system load day may still need attention if intake pressure rises sharply.
- A negative cumulative net intake may still hide short-term backlog windows.

The average system load of **6,232.77 children** shows that, across the full period, the system managed a substantial ongoing care responsibility.

### Why This KPI Matters

This KPI is important because it directly reflects operational load.

- Shelter availability

For the dashboard, this KPI should be shown as a main KPI card because it represents the most direct measure of system-wide responsibility.

## 4. KPI 2 — Net Intake Pressure

### KPI Definition

**Net Intake Pressure** measures the difference between inflow into HHS care and outflow through discharge.

### Formula

`Net Intake Pressure = Transfers into HHS Care − Discharges from HHS Care`

### Final KPI Values

| Metric | Value |
|---|---:|
| Average Net Intake Pressure | -44.74 children/day |
| Total Net Intake | -32,212 |
| Positive Net Intake Days | 238 |
| Positive Net Intake Day Rate | 33.06% |

### Business Interpretation

The average net intake pressure was **-44.74 children per day**.

This means that, on average, discharges exceeded transfers into HHS care across the full reporting period.

The total net intake was **-32,212**, which confirms that cumulative discharge activity was higher than cumulative transfer inflow.

However, the system still had **238 positive net intake days**, equal to **33.06%** of the reporting timeline.

This is important because a negative overall average does not mean that the system never experienced intake pressure.

### Why This KPI Matters

Net Intake Pressure is important because it shows whether the system is absorbing more children than it is releasing.

If transfers exceed discharges for multiple days, pressure can begin accumulating.

This can affect:

- HHS facility occupancy
- Case processing timelines
- Sponsor placement workload
- Staffing needs
- Backlog pressure

For the dashboard, this KPI should support the **Net Intake & Backlog Trends** module.

## 5. KPI 3 — Care Load Volatility Index

### KPI Definition

**Care Load Volatility Index** measures the stability of system load movement.

### Formula

`Care Load Volatility Index = Standard deviation of care load growth rate`

### Final KPI Values

| Metric | Value |
|---|---:|
| Care Load Volatility Index | 1.89 |
| System Load Standard Deviation | 2,918.46 |
| Average Absolute Growth Change | 1.24 |
| Maximum Final Pressure Score | 6.07 |

### Business Interpretation

The Care Load Volatility Index was **1.89**.

This means that daily care load growth had measurable variation across the reporting timeline.

The system load standard deviation was **2,918.46**, which indicates that total care responsibility fluctuated strongly across the full period.

The average absolute growth change was **1.24**, showing that daily load movement was usually not extreme, but the overall system still experienced meaningful variability across time.

### Why This KPI Matters

A system can be difficult to manage even when average load appears acceptable.

Volatility matters because sudden or repeated changes can create stress in:

- Facility capacity planning
- Staffing allocation
- Medical intake scheduling
- Case management distribution
- Discharge coordination
- Emergency response readiness

For the dashboard, this KPI can be shown in a pressure monitoring card or used as a trend indicator.

## 6. KPI 4 — Backlog Accumulation Rate

### KPI Definition

**Backlog Accumulation Rate** measures the share of reporting days where backlog pressure became sustained.

### Formula

`Backlog Accumulation Rate = Sustained Backlog Days / Total Reporting Days × 100`

### Final KPI Values

| Metric | Value |
|---|---:|
| Backlog Accumulation Rate | 13.33% |
| Any Backlog Pressure Days | 238 |
| Sustained Backlog Days | 96 |
| Critical Backlog Days | 27 |

### Business Interpretation

The backlog accumulation rate was **13.33%**.

This means that sustained backlog pressure appeared in a meaningful share of the reporting timeline.

There were **238 days** with any backlog pressure, **96 sustained backlog days**, and **27 critical backlog days**.

This shows that backlog pressure was not only occasional. It became sustained during some periods.

### Why This KPI Matters

Backlog pressure is important because it indicates that the system may not be clearing care load as quickly as new pressure is forming.

Sustained backlog can affect:

- Length of stay
- Shelter occupancy
- Medical and welfare support capacity
- Caseworker load
- Sponsor placement timelines
- Risk of delayed care

For the dashboard, this KPI should be connected with the **Net Intake & Backlog Trends** module.

## 7. KPI 5 — Discharge Offset Ratio

### KPI Definition

**Discharge Offset Ratio** measures whether HHS discharges are offsetting transfers into HHS care.

### Final KPI Values

The discharge offset percentage was **135%**, meaning discharges exceeded transfers by a strong margin.

The discharge surplus was **32,212**, showing that total discharges were higher than total transfer inflow.

### Why This KPI Matters

This KPI is important because it measures the system's ability to relieve load.

A ratio above 1 is generally a positive signal because outflow is greater than inflow.

However, this does not mean the system never experienced stress.

Even with a strong discharge offset ratio, there were still:

- Positive net intake days

- High stress days
- Critical pressure days

This is why discharge offset should be reviewed together with pressure classification and backlog indicators.
