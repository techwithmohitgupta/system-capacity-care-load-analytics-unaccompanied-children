### Step 5B — Create Rolling Average Pressure Metrics

This step creates rolling average pressure metrics from the daily care-load dataset.

The goal is to smooth daily fluctuations and identify short-term and medium-term pressure patterns using 7-day and 14-day rolling averages.

In this step, we will:

- Create a safe working copy of the daily pressure dataset
- Validate date order before rolling calculations
- Calculate 7-day and 14-day rolling averages for key pressure fields
- Compare current daily load against rolling baseline values
- Create rolling pressure gap indicators
- Preserve all original rows without deleting or modifying source records

These rolling metrics will support later pressure and stress identification steps, especially prolonged strain windows and high-pressure date analysis.

```python
# ============================================================
# Step 5B - Create Rolling Average Pressure Metrics
# Notebook: 05_pressure_and_stress_identification.ipynb
# ============================================================

import pandas as pd
import numpy as np

print("Step 5B - Create Rolling Average Pressure Metrics")
print("=" * 80)

# ------------------------------------------------------------
# 1. Validate Step 5A input dataset
# ------------------------------------------------------------

if "df_daily_trend" not in globals():
    raise NameError(
        "df_daily_trend not found in memory. Please run Step 5A first."
    )

df_pressure_rolling = df_daily_trend.copy()

print("Pressure rolling working copy created successfully.")
print(f"Input dataset shape: {df_pressure_rolling.shape[0]} rows × {df_pressure_rolling.shape[1]} columns")

# ------------------------------------------------------------
# 2. Validate required columns
# ------------------------------------------------------------

required_5b_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "logical_constraint_anomaly_flag"
]

missing_5b_columns = [
    col for col in required_5b_columns
    if col not in df_pressure_rolling.columns
]

print(f"Missing Step 5B input columns: {len(missing_5b_columns)}")

if missing_5b_columns:
    raise ValueError(
        f"Missing required Step 5B columns: {missing_5b_columns}"
    )

# ------------------------------------------------------------
# 3. Date validation and chronological sorting
# ------------------------------------------------------------

df_pressure_rolling["Date"] = pd.to_datetime(
    df_pressure_rolling["Date"],
    errors="coerce"
)

date_parse_errors = df_pressure_rolling["Date"].isna().sum()

if date_parse_errors > 0:
    raise ValueError(
        f"Date parse errors found: {date_parse_errors}. Please fix Date column before Step 5B."
    )

df_pressure_rolling = df_pressure_rolling.sort_values("Date").reset_index(drop=True)

is_chronological = df_pressure_rolling["Date"].is_monotonic_increasing
duplicate_dates = df_pressure_rolling["Date"].duplicated().sum()

print(f"Date parse errors: {date_parse_errors}")
print(f"Dataset chronological order: {is_chronological}")
print(f"Duplicate dates: {duplicate_dates}")

if not is_chronological:
    raise ValueError("Date column is not in chronological order after sorting.")

# ------------------------------------------------------------
# 4. Ensure numeric pressure fields
# ------------------------------------------------------------

numeric_pressure_columns = [
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score"
]

for col in numeric_pressure_columns:
    df_pressure_rolling[col] = pd.to_numeric(
        df_pressure_rolling[col],
        errors="coerce"
    )

numeric_missing_summary = pd.DataFrame({
    "Column": numeric_pressure_columns,
    "Missing Values": [df_pressure_rolling[col].isna().sum() for col in numeric_pressure_columns],
    "Minimum": [df_pressure_rolling[col].min() for col in numeric_pressure_columns],
    "Maximum": [df_pressure_rolling[col].max() for col in numeric_pressure_columns]
})

print("\nNumeric pressure field validation summary:")
display(numeric_missing_summary)

# ------------------------------------------------------------
# 5. Create 7-day rolling average metrics
# ------------------------------------------------------------

df_pressure_rolling["rolling_7d_avg_system_load"] = (
    df_pressure_rolling["total_system_load"]
    .rolling(window=7, min_periods=1)
    .mean()
)

df_pressure_rolling["rolling_7d_avg_net_intake"] = (
    df_pressure_rolling["net_daily_intake"]
    .rolling(window=7, min_periods=1)
    .mean()
)

df_pressure_rolling["rolling_7d_avg_growth_rate_pct"] = (
    df_pressure_rolling["care_load_growth_rate_pct"]
    .rolling(window=7, min_periods=1)
    .mean()
)

df_pressure_rolling["rolling_7d_avg_backlog_severity"] = (
    df_pressure_rolling["backlog_severity_score"]
    .rolling(window=7, min_periods=1)
    .mean()
)

# ------------------------------------------------------------
# 6. Create 14-day rolling average metrics
# ------------------------------------------------------------

df_pressure_rolling["rolling_14d_avg_system_load"] = (
    df_pressure_rolling["total_system_load"]
    .rolling(window=14, min_periods=1)
    .mean()
)

df_pressure_rolling["rolling_14d_avg_net_intake"] = (
    df_pressure_rolling["net_daily_intake"]
    .rolling(window=14, min_periods=1)
    .mean()
)

df_pressure_rolling["rolling_14d_avg_growth_rate_pct"] = (
    df_pressure_rolling["care_load_growth_rate_pct"]
    .rolling(window=14, min_periods=1)
    .mean()
)

df_pressure_rolling["rolling_14d_avg_backlog_severity"] = (
    df_pressure_rolling["backlog_severity_score"]
    .rolling(window=14, min_periods=1)
    .mean()
)

# ------------------------------------------------------------
# 7. Create pressure gap metrics
# ------------------------------------------------------------

df_pressure_rolling["system_load_vs_7d_avg_gap"] = (
    df_pressure_rolling["total_system_load"] -
    df_pressure_rolling["rolling_7d_avg_system_load"]
)

df_pressure_rolling["system_load_vs_14d_avg_gap"] = (
    df_pressure_rolling["total_system_load"] -
    df_pressure_rolling["rolling_14d_avg_system_load"]
)

df_pressure_rolling["net_intake_vs_7d_avg_gap"] = (
    df_pressure_rolling["net_daily_intake"] -
    df_pressure_rolling["rolling_7d_avg_net_intake"]
)

df_pressure_rolling["net_intake_vs_14d_avg_gap"] = (
    df_pressure_rolling["net_daily_intake"] -
    df_pressure_rolling["rolling_14d_avg_net_intake"]
)

# ------------------------------------------------------------
# 8. Create pressure direction indicators
# ------------------------------------------------------------

df_pressure_rolling["above_7d_system_load_avg_flag"] = (
    df_pressure_rolling["system_load_vs_7d_avg_gap"] > 0
)

df_pressure_rolling["above_14d_system_load_avg_flag"] = (
    df_pressure_rolling["system_load_vs_14d_avg_gap"] > 0
)

df_pressure_rolling["positive_7d_net_intake_pressure_flag"] = (
    df_pressure_rolling["rolling_7d_avg_net_intake"] > 0
)

df_pressure_rolling["positive_14d_net_intake_pressure_flag"] = (
    df_pressure_rolling["rolling_14d_avg_net_intake"] > 0
)

# ------------------------------------------------------------
# 9. Create rolling pressure level classification
# ------------------------------------------------------------

def classify_rolling_pressure(row):
    if (
        row["above_14d_system_load_avg_flag"] == True
        and row["positive_14d_net_intake_pressure_flag"] == True
        and row["rolling_14d_avg_backlog_severity"] >= 2
    ):
        return "Sustained Rolling Pressure"
    elif (
        row["above_7d_system_load_avg_flag"] == True
        and row["positive_7d_net_intake_pressure_flag"] == True
    ):
        return "Short-Term Rolling Pressure"
    elif row["above_7d_system_load_avg_flag"] == True:
        return "Load Above Rolling Average"
    else:
        return "Normal Rolling Pressure"

df_pressure_rolling["rolling_pressure_level"] = df_pressure_rolling.apply(
    classify_rolling_pressure,
    axis=1
)

# ------------------------------------------------------------
# 10. Round rolling metrics for readability
# ------------------------------------------------------------

rolling_numeric_columns = [
    "rolling_7d_avg_system_load",
    "rolling_7d_avg_net_intake",
    "rolling_7d_avg_growth_rate_pct",
    "rolling_7d_avg_backlog_severity",
    "rolling_14d_avg_system_load",
    "rolling_14d_avg_net_intake",
    "rolling_14d_avg_growth_rate_pct",
    "rolling_14d_avg_backlog_severity",
    "system_load_vs_7d_avg_gap",
    "system_load_vs_14d_avg_gap",
    "net_intake_vs_7d_avg_gap",
    "net_intake_vs_14d_avg_gap"
]

for col in rolling_numeric_columns:
    df_pressure_rolling[col] = df_pressure_rolling[col].round(2)

# ------------------------------------------------------------
# 11. Create rolling pressure distribution summary
# ------------------------------------------------------------

rolling_pressure_distribution = (
    df_pressure_rolling["rolling_pressure_level"]
    .value_counts()
    .reset_index()
)

rolling_pressure_distribution.columns = [
    "Rolling Pressure Level",
    "Record Count"
]

rolling_pressure_distribution["Record Percentage"] = (
    rolling_pressure_distribution["Record Count"] /
    len(df_pressure_rolling) * 100
).round(2)

# ------------------------------------------------------------
# 12. Create Step 5B validation summary
# ------------------------------------------------------------

new_5b_columns = [
    "rolling_7d_avg_system_load",
    "rolling_7d_avg_net_intake",
    "rolling_7d_avg_growth_rate_pct",
    "rolling_7d_avg_backlog_severity",
    "rolling_14d_avg_system_load",
    "rolling_14d_avg_net_intake",
    "rolling_14d_avg_growth_rate_pct",
    "rolling_14d_avg_backlog_severity",
    "system_load_vs_7d_avg_gap",
    "system_load_vs_14d_avg_gap",
    "net_intake_vs_7d_avg_gap",
    "net_intake_vs_14d_avg_gap",
    "above_7d_system_load_avg_flag",
    "above_14d_system_load_avg_flag",
    "positive_7d_net_intake_pressure_flag",
    "positive_14d_net_intake_pressure_flag",
    "rolling_pressure_level"
]

step_5b_validation_summary = pd.DataFrame({
    "Step 5B Check": [
        "Input Dataset Source",
        "Rows Checked",
        "Columns Before Step 5B",
        "Columns After Step 5B",
        "New Rolling Columns Created",
        "Missing Required Input Columns",
        "Date Parse Errors",
        "Duplicate Dates",
        "Chronological Order",
        "Missing Values in New Rolling Columns",
        "Rows Deleted or Modified"
    ],
    "Result": [
        "df_daily_trend",
        df_pressure_rolling.shape[0],
        df_daily_trend.shape[1],
        df_pressure_rolling.shape[1],
        len(new_5b_columns),
        len(missing_5b_columns),
        int(date_parse_errors),
        int(duplicate_dates),
        bool(is_chronological),
        int(df_pressure_rolling[new_5b_columns].isna().sum().sum()),
        0
    ]
})

# ------------------------------------------------------------
# 13. Preview outputs
# ------------------------------------------------------------

pressure_rolling_preview_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "rolling_7d_avg_system_load",
    "rolling_14d_avg_system_load",
    "system_load_vs_7d_avg_gap",
    "system_load_vs_14d_avg_gap",
    "rolling_7d_avg_net_intake",
    "rolling_14d_avg_net_intake",
    "rolling_pressure_level"
]

print("\nRolling pressure metrics created successfully.")
print(f"Final Step 5B dataset shape: {df_pressure_rolling.shape[0]} rows × {df_pressure_rolling.shape[1]} columns")

print("\nRolling pressure level distribution:")
display(rolling_pressure_distribution)

print("\nStep 5B validation summary:")
display(step_5b_validation_summary)

print("\nRolling pressure dataset preview:")
display(df_pressure_rolling[pressure_rolling_preview_columns].head(10))

print("\nRolling pressure dataset tail:")
display(df_pressure_rolling[pressure_rolling_preview_columns].tail(10))

print("\nStep 5B completed successfully.")
print("Rolling average pressure metrics are ready for Step 5C.")
```

## Step 5C — Create Variability / Volatility Pressure Metrics

This step creates variability and volatility-based pressure metrics to identify unstable care load patterns across the system.

In this step, we calculate rolling variability indicators for key pressure fields such as total system load, net daily intake, and care load growth rate. These metrics help detect short-term fluctuations, unstable operating windows, and early signs of system pressure that may not be visible through simple averages alone.

The output from this step will support later pressure and stress identification by adding volatility-focused features to the rolling pressure dataset.

```python
# ============================================================
# Step 5C - Create Variability / Volatility Pressure Metrics
# ============================================================

import pandas as pd
import numpy as np

print("Step 5C - Create Variability / Volatility Pressure Metrics")
print("=" * 75)

# ------------------------------------------------------------
# 1. Select safest available input dataset
# ------------------------------------------------------------

if "df_pressure_rolling" in globals():
    df_pressure_volatility = df_pressure_rolling.copy()
    input_dataset_source = "df_pressure_rolling"
elif "df_daily_trend_high_load" in globals():
    df_pressure_volatility = df_daily_trend_high_load.copy()
    input_dataset_source = "df_daily_trend_high_load"
elif "df_daily_trend" in globals():
    df_pressure_volatility = df_daily_trend.copy()
    input_dataset_source = "df_daily_trend"
else:
    raise NameError(
        "No valid Step 5C input dataset found. Expected one of: "
        "'df_pressure_rolling', 'df_daily_trend_high_load', or 'df_daily_trend'. "
        "Please run Step 5A and Step 5B first."
    )

print(f"Pressure volatility working copy created successfully.")
print(f"Input dataset source: {input_dataset_source}")
print(f"Input dataset shape: {df_pressure_volatility.shape[0]} rows × {df_pressure_volatility.shape[1]} columns")

# ------------------------------------------------------------
# 2. Validate required columns
# ------------------------------------------------------------

required_5c_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score"
]

missing_5c_columns = [
    col for col in required_5c_columns
    if col not in df_pressure_volatility.columns
]

print(f"Missing Step 5C input columns: {len(missing_5c_columns)}")

if missing_5c_columns:
    raise ValueError(f"Missing required Step 5C columns: {missing_5c_columns}")

# ------------------------------------------------------------
# 3. Validate and sort Date column
# ------------------------------------------------------------

df_pressure_volatility["Date"] = pd.to_datetime(
    df_pressure_volatility["Date"],
    errors="coerce"
)

date_parse_errors_5c = int(df_pressure_volatility["Date"].isna().sum())

if date_parse_errors_5c > 0:
    raise ValueError(
        f"Date parse errors found in Step 5C: {date_parse_errors_5c}. "
        "Please fix Date values before continuing."
    )

df_pressure_volatility = (
    df_pressure_volatility
    .sort_values("Date")
    .reset_index(drop=True)
)

duplicate_dates_5c = int(df_pressure_volatility["Date"].duplicated().sum())
is_chronological_5c = bool(df_pressure_volatility["Date"].is_monotonic_increasing)

print(f"Date parse errors: {date_parse_errors_5c}")
print(f"Dataset chronological order: {is_chronological_5c}")
print(f"Duplicate dates: {duplicate_dates_5c}")

# ------------------------------------------------------------
# 4. Validate numeric fields
# ------------------------------------------------------------

numeric_pressure_fields_5c = [
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score"
]

for col in numeric_pressure_fields_5c:
    df_pressure_volatility[col] = pd.to_numeric(
        df_pressure_volatility[col],
        errors="coerce"
    )

numeric_validation_5c = pd.DataFrame({
    "Column": numeric_pressure_fields_5c,
    "Missing Values": [int(df_pressure_volatility[col].isna().sum()) for col in numeric_pressure_fields_5c],
    "Minimum": [df_pressure_volatility[col].min() for col in numeric_pressure_fields_5c],
    "Maximum": [df_pressure_volatility[col].max() for col in numeric_pressure_fields_5c]
})

print("\nNumeric volatility field validation summary:")
display(numeric_validation_5c)

# ------------------------------------------------------------
# 5. Create rolling volatility metrics
# ------------------------------------------------------------

# 7-day volatility metrics
df_pressure_volatility["rolling_7d_std_system_load"] = (
    df_pressure_volatility["total_system_load"]
    .rolling(window=7, min_periods=1)
    .std()
    .fillna(0)
)

df_pressure_volatility["rolling_7d_std_net_intake"] = (
    df_pressure_volatility["net_daily_intake"]
    .rolling(window=7, min_periods=1)
    .std()
    .fillna(0)
)

df_pressure_volatility["rolling_7d_std_growth_rate"] = (
    df_pressure_volatility["care_load_growth_rate_pct"]
    .rolling(window=7, min_periods=1)
    .std()
    .fillna(0)
)

# 14-day volatility metrics
df_pressure_volatility["rolling_14d_std_system_load"] = (
    df_pressure_volatility["total_system_load"]
    .rolling(window=14, min_periods=1)
    .std()
    .fillna(0)
)

df_pressure_volatility["rolling_14d_std_net_intake"] = (
    df_pressure_volatility["net_daily_intake"]
    .rolling(window=14, min_periods=1)
    .std()
    .fillna(0)
)

df_pressure_volatility["rolling_14d_std_growth_rate"] = (
    df_pressure_volatility["care_load_growth_rate_pct"]
    .rolling(window=14, min_periods=1)
    .std()
    .fillna(0)
)

# ------------------------------------------------------------
# 6. Create rolling range / swing metrics
# ------------------------------------------------------------

df_pressure_volatility["rolling_7d_system_load_range"] = (
    df_pressure_volatility["total_system_load"]
    .rolling(window=7, min_periods=1)
    .max()
    -
    df_pressure_volatility["total_system_load"]
    .rolling(window=7, min_periods=1)
    .min()
)

df_pressure_volatility["rolling_14d_system_load_range"] = (
    df_pressure_volatility["total_system_load"]
    .rolling(window=14, min_periods=1)
    .max()
    -
    df_pressure_volatility["total_system_load"]
    .rolling(window=14, min_periods=1)
    .min()
)

df_pressure_volatility["rolling_7d_net_intake_range"] = (
    df_pressure_volatility["net_daily_intake"]
    .rolling(window=7, min_periods=1)
    .max()
    -
    df_pressure_volatility["net_daily_intake"]
    .rolling(window=7, min_periods=1)
    .min()
)

df_pressure_volatility["rolling_14d_net_intake_range"] = (
    df_pressure_volatility["net_daily_intake"]
    .rolling(window=14, min_periods=1)
    .max()
    -
    df_pressure_volatility["net_daily_intake"]
    .rolling(window=14, min_periods=1)
    .min()
)

# ------------------------------------------------------------
# 7. Create volatility intensity score
# ------------------------------------------------------------

load_volatility_threshold_75 = df_pressure_volatility["rolling_14d_std_system_load"].quantile(0.75)
load_volatility_threshold_90 = df_pressure_volatility["rolling_14d_std_system_load"].quantile(0.90)

intake_volatility_threshold_75 = df_pressure_volatility["rolling_14d_std_net_intake"].quantile(0.75)
intake_volatility_threshold_90 = df_pressure_volatility["rolling_14d_std_net_intake"].quantile(0.90)

growth_volatility_threshold_75 = df_pressure_volatility["rolling_14d_std_growth_rate"].quantile(0.75)
growth_volatility_threshold_90 = df_pressure_volatility["rolling_14d_std_growth_rate"].quantile(0.90)

df_pressure_volatility["system_load_volatility_flag"] = np.select(
    [
        df_pressure_volatility["rolling_14d_std_system_load"] >= load_volatility_threshold_90,
        df_pressure_volatility["rolling_14d_std_system_load"] >= load_volatility_threshold_75
    ],
    [
        "High Load Volatility",
        "Moderate Load Volatility"
    ],
    default="Stable Load Volatility"
)

df_pressure_volatility["net_intake_volatility_flag"] = np.select(
    [
        df_pressure_volatility["rolling_14d_std_net_intake"] >= intake_volatility_threshold_90,
        df_pressure_volatility["rolling_14d_std_net_intake"] >= intake_volatility_threshold_75
    ],
    [
        "High Intake Volatility",
        "Moderate Intake Volatility"
    ],
    default="Stable Intake Volatility"
)

df_pressure_volatility["growth_rate_volatility_flag"] = np.select(
    [
        df_pressure_volatility["rolling_14d_std_growth_rate"] >= growth_volatility_threshold_90,
        df_pressure_volatility["rolling_14d_std_growth_rate"] >= growth_volatility_threshold_75
    ],
    [
        "High Growth Volatility",
        "Moderate Growth Volatility"
    ],
    default="Stable Growth Volatility"
)

# Composite numeric volatility score
df_pressure_volatility["volatility_pressure_score"] = (
    np.where(df_pressure_volatility["rolling_14d_std_system_load"] >= load_volatility_threshold_90, 2,
    np.where(df_pressure_volatility["rolling_14d_std_system_load"] >= load_volatility_threshold_75, 1, 0))
    +
    np.where(df_pressure_volatility["rolling_14d_std_net_intake"] >= intake_volatility_threshold_90, 2,
    np.where(df_pressure_volatility["rolling_14d_std_net_intake"] >= intake_volatility_threshold_75, 1, 0))
    +
    np.where(df_pressure_volatility["rolling_14d_std_growth_rate"] >= growth_volatility_threshold_90, 2,
    np.where(df_pressure_volatility["rolling_14d_std_growth_rate"] >= growth_volatility_threshold_75, 1, 0))
    +
    np.where(df_pressure_volatility["backlog_severity_score"] >= 2, 1, 0)
)

df_pressure_volatility["volatility_pressure_level"] = np.select(
    [
        df_pressure_volatility["volatility_pressure_score"] >= 5,
        df_pressure_volatility["volatility_pressure_score"] >= 3,
        df_pressure_volatility["volatility_pressure_score"] >= 1
    ],
    [
        "High Volatility Pressure",
        "Moderate Volatility Pressure",
        "Low Volatility Pressure"
    ],
    default="Stable Pressure"
)

print("\nVolatility pressure metrics created successfully.")

# ------------------------------------------------------------
# 8. Create volatility distribution summary
# ------------------------------------------------------------

volatility_pressure_distribution = (
    df_pressure_volatility["volatility_pressure_level"]
    .value_counts()
    .reset_index()
)

volatility_pressure_distribution.columns = [
    "Volatility Pressure Level",
    "Record Count"
]

volatility_pressure_distribution["Record Percentage"] = (
    volatility_pressure_distribution["Record Count"]
    / len(df_pressure_volatility)
    * 100
).round(2)

volatility_order = {
    "Stable Pressure": 1,
    "Low Volatility Pressure": 2,
    "Moderate Volatility Pressure": 3,
    "High Volatility Pressure": 4
}

volatility_pressure_distribution["Sort Order"] = (
    volatility_pressure_distribution["Volatility Pressure Level"]
    .map(volatility_order)
)

volatility_pressure_distribution = (
    volatility_pressure_distribution
    .sort_values("Sort Order")
    .drop(columns="Sort Order")
    .reset_index(drop=True)
)

print("\nVolatility pressure level distribution:")
display(volatility_pressure_distribution)

# ------------------------------------------------------------
# 9. Create Step 5C validation summary
# ------------------------------------------------------------

new_5c_columns = [
    "rolling_7d_std_system_load",
    "rolling_7d_std_net_intake",
    "rolling_7d_std_growth_rate",
    "rolling_14d_std_system_load",
    "rolling_14d_std_net_intake",
    "rolling_14d_std_growth_rate",
    "rolling_7d_system_load_range",
    "rolling_14d_system_load_range",
    "rolling_7d_net_intake_range",
    "rolling_14d_net_intake_range",
    "system_load_volatility_flag",
    "net_intake_volatility_flag",
    "growth_rate_volatility_flag",
    "volatility_pressure_score",
    "volatility_pressure_level"
]

missing_new_5c_columns = [
    col for col in new_5c_columns
    if col not in df_pressure_volatility.columns
]

missing_values_new_5c = int(
    df_pressure_volatility[new_5c_columns].isna().sum().sum()
) if not missing_new_5c_columns else None

step_5c_validation_summary = pd.DataFrame({
    "Step 5C Check": [
        "Input Dataset Source",
        "Rows Checked",
        "Columns Before Step 5C",
        "Columns After Step 5C",
        "New Volatility Columns Created",
        "Missing Required Input Columns",
        "Missing New Volatility Columns",
        "Missing Values in New Volatility Columns",
        "Date Parse Errors",
        "Duplicate Dates",
        "Chronological Order",
        "Rows Deleted or Modified"
    ],
    "Result": [
        input_dataset_source,
        df_pressure_volatility.shape[0],
        df_pressure_rolling.shape[1] if "df_pressure_rolling" in globals() else "Not Available",
        df_pressure_volatility.shape[1],
        len(new_5c_columns),
        len(missing_5c_columns),
        len(missing_new_5c_columns),
        missing_values_new_5c,
        date_parse_errors_5c,
        duplicate_dates_5c,
        is_chronological_5c,
        0
    ]
})

print("\nStep 5C validation summary:")
display(step_5c_validation_summary)

# ------------------------------------------------------------
# 10. Preview final volatility pressure dataset
# ------------------------------------------------------------

pressure_volatility_preview_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "rolling_7d_std_system_load",
    "rolling_14d_std_system_load",
    "rolling_7d_std_net_intake",
    "rolling_14d_std_net_intake",
    "rolling_7d_system_load_range",
    "rolling_14d_system_load_range",
    "system_load_volatility_flag",
    "net_intake_volatility_flag",
    "growth_rate_volatility_flag",
    "volatility_pressure_score",
    "volatility_pressure_level"
]

pressure_volatility_preview_columns = [
    col for col in pressure_volatility_preview_columns
    if col in df_pressure_volatility.columns
]

print("\nVolatility pressure dataset preview:")
display(df_pressure_volatility[pressure_volatility_preview_columns].head(10))

print("\nVolatility pressure dataset tail:")
display(df_pressure_volatility[pressure_volatility_preview_columns].tail(10))

print("\nStep 5C completed successfully.")
print("Variability / volatility pressure metrics are ready for Step 5D.")
```

## Step 5D — Create Pressure Spike / Sudden Change Indicators

This step identifies sudden pressure spikes and sharp day-over-day changes in the care load system.

The purpose of this step is to detect short-term stress signals that may not be fully visible through rolling averages or volatility metrics alone.

In this step, we will:

- Create daily change metrics for system load, net intake, HHS care load, and CBP custody load
- Calculate percentage change indicators
- Flag sudden system load spikes
- Flag sudden intake pressure changes
- Classify spike intensity levels
- Create a combined pressure spike score
- Validate that no rows are deleted or modified
- Prepare the spike indicator dataset for the next step

```python
# ------------------------------------------------------------
# Step 5D — Create Pressure Spike / Sudden Change Indicators
# ------------------------------------------------------------

import pandas as pd
import numpy as np

print("Step 5D - Create Pressure Spike / Sudden Change Indicators")
print("=" * 80)

# ------------------------------------------------------------
# 1. Select safest available input dataset
# ------------------------------------------------------------

if "df_pressure_volatility" in globals():
    df_pressure_spike = df_pressure_volatility.copy()
    input_dataset_source = "df_pressure_volatility"
elif "df_pressure_rolling" in globals():
    df_pressure_spike = df_pressure_rolling.copy()
    input_dataset_source = "df_pressure_rolling"
elif "df_daily_trend" in globals():
    df_pressure_spike = df_daily_trend.copy()
    input_dataset_source = "df_daily_trend"
else:
    raise NameError(
        "No valid pressure dataset found. Expected one of: "
        "'df_pressure_volatility', 'df_pressure_rolling', or 'df_daily_trend'. "
        "Please run previous steps first."
    )

rows_before_5d = df_pressure_spike.shape[0]
cols_before_5d = df_pressure_spike.shape[1]

print(f"Pressure spike working copy created successfully.")
print(f"Input dataset source: {input_dataset_source}")
print(f"Input dataset shape: {rows_before_5d} rows × {cols_before_5d} columns")

# ------------------------------------------------------------
# 2. Validate required input columns
# ------------------------------------------------------------

required_5d_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score"
]

optional_5d_columns = [
    "Children in CBP custody",
    "Children in HHS Care",
    "Children transferred out of CBP custody",
    "Children discharged from HHS Care"
]

missing_5d_columns = [
    col for col in required_5d_columns
    if col not in df_pressure_spike.columns
]

print(f"Missing Step 5D required input columns: {len(missing_5d_columns)}")

if len(missing_5d_columns) > 0:
    print("Missing columns:")
    for col in missing_5d_columns:
        print(f"- {col}")
    raise KeyError("Step 5D cannot continue because required input columns are missing.")

# ------------------------------------------------------------
# 3. Date validation and chronological sorting
# ------------------------------------------------------------

df_pressure_spike["Date"] = pd.to_datetime(df_pressure_spike["Date"], errors="coerce")

date_parse_errors_5d = int(df_pressure_spike["Date"].isna().sum())

if date_parse_errors_5d > 0:
    raise ValueError(
        f"Step 5D found {date_parse_errors_5d} invalid Date values. "
        "Please fix Date column before continuing."
    )

df_pressure_spike = df_pressure_spike.sort_values("Date").reset_index(drop=True)

chronological_order_5d = df_pressure_spike["Date"].is_monotonic_increasing
duplicate_dates_5d = int(df_pressure_spike["Date"].duplicated().sum())

print(f"Date parse errors: {date_parse_errors_5d}")
print(f"Dataset chronological order: {chronological_order_5d}")
print(f"Duplicate dates: {duplicate_dates_5d}")

# ------------------------------------------------------------
# 4. Create absolute day-over-day change metrics
# ------------------------------------------------------------

df_pressure_spike["daily_system_load_change"] = (
    df_pressure_spike["total_system_load"].diff().fillna(0)
)

df_pressure_spike["daily_net_intake_change"] = (
    df_pressure_spike["net_daily_intake"].diff().fillna(0)
)

df_pressure_spike["daily_growth_rate_change"] = (
    df_pressure_spike["care_load_growth_rate_pct"].diff().fillna(0)
)

df_pressure_spike["daily_backlog_severity_change"] = (
    df_pressure_spike["backlog_severity_score"].diff().fillna(0)
)

# Optional operational change metrics
if "Children in CBP custody" in df_pressure_spike.columns:
    df_pressure_spike["daily_cbp_custody_change"] = (
        df_pressure_spike["Children in CBP custody"].diff().fillna(0)
    )
else:
    df_pressure_spike["daily_cbp_custody_change"] = 0

if "Children in HHS Care" in df_pressure_spike.columns:
    df_pressure_spike["daily_hhs_care_change"] = (
        df_pressure_spike["Children in HHS Care"].diff().fillna(0)
    )
else:
    df_pressure_spike["daily_hhs_care_change"] = 0

# ------------------------------------------------------------
# 5. Create percentage change indicators
# ------------------------------------------------------------

df_pressure_spike["system_load_pct_change"] = (
    df_pressure_spike["total_system_load"]
    .pct_change()
    .replace([np.inf, -np.inf], 0)
    .fillna(0) * 100
)

df_pressure_spike["net_intake_pct_change"] = (
    df_pressure_spike["net_daily_intake"]
    .replace(0, np.nan)
    .pct_change()
    .replace([np.inf, -np.inf], 0)
    .fillna(0) * 100
)

df_pressure_spike["cbp_custody_pct_change"] = (
    df_pressure_spike["Children in CBP custody"]
    .pct_change()
    .replace([np.inf, -np.inf], 0)
    .fillna(0) * 100
    if "Children in CBP custody" in df_pressure_spike.columns
    else 0
)

df_pressure_spike["hhs_care_pct_change"] = (
    df_pressure_spike["Children in HHS Care"]
    .pct_change()
    .replace([np.inf, -np.inf], 0)
    .fillna(0) * 100
    if "Children in HHS Care" in df_pressure_spike.columns
    else 0
)

# ------------------------------------------------------------
# 6. Create spike thresholds using distribution-based logic
# ------------------------------------------------------------

system_load_change_abs = df_pressure_spike["daily_system_load_change"].abs()
net_intake_change_abs = df_pressure_spike["daily_net_intake_change"].abs()
growth_rate_change_abs = df_pressure_spike["daily_growth_rate_change"].abs()

system_load_spike_threshold = system_load_change_abs.quantile(0.90)
major_system_load_spike_threshold = system_load_change_abs.quantile(0.95)

net_intake_spike_threshold = net_intake_change_abs.quantile(0.90)
major_net_intake_spike_threshold = net_intake_change_abs.quantile(0.95)

growth_rate_spike_threshold = growth_rate_change_abs.quantile(0.90)
major_growth_rate_spike_threshold = growth_rate_change_abs.quantile(0.95)

print("\nSpike threshold values:")
print(f"90th percentile system load change: {system_load_spike_threshold:.2f}")
print(f"95th percentile system load change: {major_system_load_spike_threshold:.2f}")
print(f"90th percentile net intake change: {net_intake_spike_threshold:.2f}")
print(f"95th percentile net intake change: {major_net_intake_spike_threshold:.2f}")
print(f"90th percentile growth rate change: {growth_rate_spike_threshold:.2f}")
print(f"95th percentile growth rate change: {major_growth_rate_spike_threshold:.2f}")

# ------------------------------------------------------------
# 7. Create spike flags
# ------------------------------------------------------------

df_pressure_spike["system_load_spike_flag"] = (
    system_load_change_abs >= system_load_spike_threshold
)

df_pressure_spike["major_system_load_spike_flag"] = (
    system_load_change_abs >= major_system_load_spike_threshold
)

df_pressure_spike["net_intake_spike_flag"] = (
    net_intake_change_abs >= net_intake_spike_threshold
)

df_pressure_spike["major_net_intake_spike_flag"] = (
    net_intake_change_abs >= major_net_intake_spike_threshold
)

df_pressure_spike["growth_rate_spike_flag"] = (
    growth_rate_change_abs >= growth_rate_spike_threshold
)

df_pressure_spike["major_growth_rate_spike_flag"] = (
    growth_rate_change_abs >= major_growth_rate_spike_threshold
)

df_pressure_spike["backlog_severity_jump_flag"] = (
    df_pressure_spike["daily_backlog_severity_change"] > 0
)

# ------------------------------------------------------------
# 8. Create directional spike interpretation
# ------------------------------------------------------------

def classify_system_load_direction(change_value):
    if change_value > 0:
        return "System Load Increase"
    elif change_value < 0:
        return "System Load Decrease"
    else:
        return "No System Load Change"

def classify_net_intake_direction(change_value):
    if change_value > 0:
        return "Net Intake Increase"
    elif change_value < 0:
        return "Net Intake Decrease"
    else:
        return "No Net Intake Change"

df_pressure_spike["system_load_change_direction"] = (
    df_pressure_spike["daily_system_load_change"].apply(classify_system_load_direction)
)

df_pressure_spike["net_intake_change_direction"] = (
    df_pressure_spike["daily_net_intake_change"].apply(classify_net_intake_direction)
)

# ------------------------------------------------------------
# 9. Create combined spike score
# ------------------------------------------------------------

df_pressure_spike["pressure_spike_score"] = (
    df_pressure_spike["system_load_spike_flag"].astype(int)
    + df_pressure_spike["major_system_load_spike_flag"].astype(int)
    + df_pressure_spike["net_intake_spike_flag"].astype(int)
    + df_pressure_spike["major_net_intake_spike_flag"].astype(int)
    + df_pressure_spike["growth_rate_spike_flag"].astype(int)
    + df_pressure_spike["major_growth_rate_spike_flag"].astype(int)
    + df_pressure_spike["backlog_severity_jump_flag"].astype(int)
)

# ------------------------------------------------------------
# 10. Classify pressure spike level
# ------------------------------------------------------------

def classify_pressure_spike_level(score):
    if score >= 5:
        return "Critical Pressure Spike"
    elif score >= 3:
        return "High Pressure Spike"
    elif score >= 1:
        return "Moderate Pressure Spike"
    else:
        return "No Major Spike"

df_pressure_spike["pressure_spike_level"] = (
    df_pressure_spike["pressure_spike_score"].apply(classify_pressure_spike_level)
)

# ------------------------------------------------------------
# 11. Create spike risk category
# ------------------------------------------------------------

def classify_spike_risk(row):
    if (
        row["major_system_load_spike_flag"]
        and row["system_load_change_direction"] == "System Load Increase"
    ):
        return "Major Load Increase Risk"
    elif (
        row["major_net_intake_spike_flag"]
        and row["net_intake_change_direction"] == "Net Intake Increase"
    ):
        return "Major Intake Increase Risk"
    elif row["backlog_severity_jump_flag"]:
        return "Backlog Severity Escalation"
    elif row["pressure_spike_score"] >= 3:
        return "Multi-Factor Spike Risk"
    elif row["pressure_spike_score"] >= 1:
        return "Moderate Spike Watch"
    else:
        return "No Spike Risk"

df_pressure_spike["pressure_spike_risk_category"] = (
    df_pressure_spike.apply(classify_spike_risk, axis=1)
)

# ------------------------------------------------------------
# 12. Create spike distribution summaries
# ------------------------------------------------------------

pressure_spike_level_distribution = (
    df_pressure_spike["pressure_spike_level"]
    .value_counts()
    .rename_axis("Pressure Spike Level")
    .reset_index(name="Record Count")
)

pressure_spike_level_distribution["Record Percentage"] = (
    pressure_spike_level_distribution["Record Count"] / len(df_pressure_spike) * 100
).round(2)

spike_risk_category_distribution = (
    df_pressure_spike["pressure_spike_risk_category"]
    .value_counts()
    .rename_axis("Pressure Spike Risk Category")
    .reset_index(name="Record Count")
)

spike_risk_category_distribution["Record Percentage"] = (
    spike_risk_category_distribution["Record Count"] / len(df_pressure_spike) * 100
).round(2)

# ------------------------------------------------------------
# 13. Create validation summary
# ------------------------------------------------------------

new_5d_columns = [
    "daily_system_load_change",
    "daily_net_intake_change",
    "daily_growth_rate_change",
    "daily_backlog_severity_change",
    "daily_cbp_custody_change",
    "daily_hhs_care_change",
    "system_load_pct_change",
    "net_intake_pct_change",
    "cbp_custody_pct_change",
    "hhs_care_pct_change",
    "system_load_spike_flag",
    "major_system_load_spike_flag",
    "net_intake_spike_flag",
    "major_net_intake_spike_flag",
    "growth_rate_spike_flag",
    "major_growth_rate_spike_flag",
    "backlog_severity_jump_flag",
    "system_load_change_direction",
    "net_intake_change_direction",
    "pressure_spike_score",
    "pressure_spike_level",
    "pressure_spike_risk_category"
]

missing_new_5d_columns = [
    col for col in new_5d_columns
    if col not in df_pressure_spike.columns
]

missing_values_new_5d_columns = int(
    df_pressure_spike[new_5d_columns].isna().sum().sum()
)

rows_after_5d = df_pressure_spike.shape[0]
cols_after_5d = df_pressure_spike.shape[1]

rows_deleted_or_modified_5d = rows_before_5d - rows_after_5d

step_5d_validation_summary = pd.DataFrame({
    "Step 5D Check": [
        "Input Dataset Source",
        "Rows Checked",
        "Columns Before Step 5D",
        "Columns After Step 5D",
        "New Spike Columns Created",
        "Missing Required Input Columns",
        "Missing New Spike Columns",
        "Missing Values in New Spike Columns",
        "Date Parse Errors",
        "Duplicate Dates",
        "Chronological Order",
        "Rows Deleted or Modified"
    ],
    "Result": [
        input_dataset_source,
        rows_after_5d,
        cols_before_5d,
        cols_after_5d,
        len(new_5d_columns),
        len(missing_5d_columns),
        len(missing_new_5d_columns),
        missing_values_new_5d_columns,
        date_parse_errors_5d,
        duplicate_dates_5d,
        chronological_order_5d,
        rows_deleted_or_modified_5d
    ]
})

# ------------------------------------------------------------
# 14. Preview columns
# ------------------------------------------------------------

pressure_spike_preview_columns = [
    "Date",
    "total_system_load",
    "daily_system_load_change",
    "system_load_pct_change",
    "net_daily_intake",
    "daily_net_intake_change",
    "net_intake_pct_change",
    "care_load_growth_rate_pct",
    "daily_growth_rate_change",
    "backlog_severity_score",
    "pressure_spike_score",
    "pressure_spike_level",
    "pressure_spike_risk_category"
]

pressure_spike_preview_columns = [
    col for col in pressure_spike_preview_columns
    if col in df_pressure_spike.columns
]

# ------------------------------------------------------------
# 15. Final display outputs
# ------------------------------------------------------------

print("\nPressure spike indicators created successfully.")
print(f"Final Step 5D dataset shape: {rows_after_5d} rows × {cols_after_5d} columns")

print("\nPressure spike level distribution:")
display(pressure_spike_level_distribution)

print("\nPressure spike risk category distribution:")
display(spike_risk_category_distribution)

print("\nStep 5D validation summary:")
display(step_5d_validation_summary)

print("\nPressure spike dataset preview:")
display(df_pressure_spike[pressure_spike_preview_columns].head(10))

print("\nPressure spike dataset tail:")
display(df_pressure_spike[pressure_spike_preview_columns].tail(10))

print("\nStep 5D completed successfully.")
print("Pressure spike / sudden change indicators are ready for Step 5E.")
```

## Step 5E — Detect Prolonged Strain / High-Pressure Windows

In this step, we identify prolonged pressure and strain windows across the daily care load timeline.

This step uses the pressure-enriched dataset created in the previous steps and detects continuous high-pressure periods based on multiple operational signals, including:

- Rolling pressure level
- Volatility pressure level
- Pressure spike level
- Backlog severity
- High-load day indicators
- Logical anomaly pressure

The goal of this step is to move beyond single-day spikes and identify sustained operational stress windows where the care system remained under pressure for multiple consecutive reporting dates.

This output will support the next analysis stages by helping identify:

- Prolonged strain periods
- High-pressure date windows
- System pressure duration
- Peak pressure windows
- Critical operational stress periods

```python
# ============================================================
# Step 5E — Detect Prolonged Strain / High-Pressure Windows
# ============================================================

import pandas as pd
import numpy as np

print("Step 5E - Detect Prolonged Strain / High-Pressure Windows")
print("=" * 80)

# ------------------------------------------------------------
# 1. Select safest available input dataframe
# ------------------------------------------------------------

input_source_name = None

if "df_pressure_spike" in globals():
    df_pressure_windows_base = df_pressure_spike.copy()
    input_source_name = "df_pressure_spike"
elif "df_pressure_volatility" in globals():
    df_pressure_windows_base = df_pressure_volatility.copy()
    input_source_name = "df_pressure_volatility"
elif "df_pressure_rolling" in globals():
    df_pressure_windows_base = df_pressure_rolling.copy()
    input_source_name = "df_pressure_rolling"
elif "df_daily_trend_high_load" in globals():
    df_pressure_windows_base = df_daily_trend_high_load.copy()
    input_source_name = "df_daily_trend_high_load"
elif "df_daily_trend" in globals():
    df_pressure_windows_base = df_daily_trend.copy()
    input_source_name = "df_daily_trend"
else:
    raise NameError(
        "No valid input dataframe found for Step 5E. "
        "Please run Step 5A to Step 5D first."
    )

print(f"Pressure window working copy created successfully.")
print(f"Input dataset source: {input_source_name}")
print(f"Input dataset shape: {df_pressure_windows_base.shape[0]} rows × {df_pressure_windows_base.shape[1]} columns")

# ------------------------------------------------------------
# 2. Validate date column
# ------------------------------------------------------------

if "Date" not in df_pressure_windows_base.columns:
    raise KeyError("Required column missing: Date")

df_pressure_windows_base["Date"] = pd.to_datetime(
    df_pressure_windows_base["Date"],
    errors="coerce"
)

date_parse_errors = int(df_pressure_windows_base["Date"].isna().sum())

if date_parse_errors > 0:
    raise ValueError(
        f"Date parsing failed for {date_parse_errors} rows. "
        "Please fix Date column before Step 5E."
    )

df_pressure_windows_base = (
    df_pressure_windows_base
    .sort_values("Date")
    .reset_index(drop=True)
)

chronological_order = bool(df_pressure_windows_base["Date"].is_monotonic_increasing)
duplicate_dates = int(df_pressure_windows_base["Date"].duplicated().sum())

print(f"Date parse errors: {date_parse_errors}")
print(f"Dataset chronological order: {chronological_order}")
print(f"Duplicate dates: {duplicate_dates}")

# ------------------------------------------------------------
# 3. Required base metric validation
# ------------------------------------------------------------

base_required_columns = [
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score"
]

missing_base_columns = [
    col for col in base_required_columns
    if col not in df_pressure_windows_base.columns
]

print(f"Missing Step 5E base input columns: {len(missing_base_columns)}")

if missing_base_columns:
    print("Missing columns:")
    for col in missing_base_columns:
        print(f"- {col}")
    raise KeyError(
        "Step 5E cannot continue because required base pressure columns are missing."
    )

# ------------------------------------------------------------
# 4. Helper functions
# ------------------------------------------------------------

def safe_numeric(series):
    return pd.to_numeric(series, errors="coerce")


def safe_flag_from_text(series, positive_keywords):
    """
    Converts text/category columns into 1/0 flags based on keyword matches.
    """
    text_series = series.fillna("").astype(str).str.lower()
    pattern = "|".join([str(keyword).lower() for keyword in positive_keywords])
    return text_series.str.contains(pattern, regex=True).astype(int)


def safe_binary_column(df, column_name, positive_keywords=None, default_value=0):
    """
    Safely creates a binary signal from an existing column.
    Works for numeric, bool, or text columns.
    """
    if column_name not in df.columns:
        return pd.Series(default_value, index=df.index)

    series = df[column_name]

    if pd.api.types.is_bool_dtype(series):
        return series.fillna(False).astype(int)

    if pd.api.types.is_numeric_dtype(series):
        return (pd.to_numeric(series, errors="coerce").fillna(0) > 0).astype(int)

    if positive_keywords is None:
        positive_keywords = ["high", "critical", "sustained", "pressure", "risk", "spike", "elevated"]

    return safe_flag_from_text(series, positive_keywords)


# ------------------------------------------------------------
# 5. Create individual pressure signals safely
# ------------------------------------------------------------

df_pressure_windows = df_pressure_windows_base.copy()

# Rolling pressure signal
if "rolling_pressure_level" in df_pressure_windows.columns:
    df_pressure_windows["rolling_pressure_signal"] = safe_flag_from_text(
        df_pressure_windows["rolling_pressure_level"],
        ["short-term rolling pressure", "sustained rolling pressure", "load above rolling average"]
    )
else:
    df_pressure_windows["rolling_pressure_signal"] = 0

# Volatility pressure signal
if "volatility_pressure_level" in df_pressure_windows.columns:
    df_pressure_windows["volatility_pressure_signal"] = safe_flag_from_text(
        df_pressure_windows["volatility_pressure_level"],
        ["moderate volatility pressure", "high volatility pressure"]
    )
else:
    df_pressure_windows["volatility_pressure_signal"] = 0

# Pressure spike signal
if "pressure_spike_level" in df_pressure_windows.columns:
    df_pressure_windows["pressure_spike_signal"] = safe_flag_from_text(
        df_pressure_windows["pressure_spike_level"],
        ["moderate pressure spike", "high pressure spike", "critical pressure spike"]
    )
else:
    df_pressure_windows["pressure_spike_signal"] = 0

# Pressure spike risk signal
if "pressure_spike_risk_category" in df_pressure_windows.columns:
    df_pressure_windows["pressure_spike_risk_signal"] = safe_flag_from_text(
        df_pressure_windows["pressure_spike_risk_category"],
        [
            "backlog severity escalation",
            "moderate spike watch",
            "multi-factor spike risk",
            "major intake increase risk",
            "major load increase risk"
        ]
    )
else:
    df_pressure_windows["pressure_spike_risk_signal"] = 0

# High-load signal from Notebook 04 / Step 4G
high_load_signal_columns = [
    "high_load_day_flag",
    "elevated_or_high_load_day_flag",
    "critical_load_day_flag",
    "sustained_high_load_flag"
]

existing_high_load_signal_columns = [
    col for col in high_load_signal_columns
    if col in df_pressure_windows.columns
]

if existing_high_load_signal_columns:
    high_load_signal_frame = pd.DataFrame(index=df_pressure_windows.index)

    for col in existing_high_load_signal_columns:
        high_load_signal_frame[col] = safe_binary_column(df_pressure_windows, col)

    df_pressure_windows["high_load_signal"] = (
        high_load_signal_frame.sum(axis=1) > 0
    ).astype(int)
else:
    df_pressure_windows["high_load_signal"] = 0

# Backlog pressure signal
df_pressure_windows["backlog_pressure_signal"] = (
    safe_numeric(df_pressure_windows["backlog_severity_score"]).fillna(0) >= 2
).astype(int)

# Logical anomaly signal
if "logical_constraint_anomaly_flag" in df_pressure_windows.columns:
    df_pressure_windows["logical_anomaly_signal"] = safe_binary_column(
        df_pressure_windows,
        "logical_constraint_anomaly_flag"
    )
elif "logical_anomaly_days" in df_pressure_windows.columns:
    df_pressure_windows["logical_anomaly_signal"] = (
        safe_numeric(df_pressure_windows["logical_anomaly_days"]).fillna(0) > 0
    ).astype(int)
else:
    df_pressure_windows["logical_anomaly_signal"] = 0

# Positive intake / intake pressure signal
if "positive_net_intake_flag" in df_pressure_windows.columns:
    df_pressure_windows["intake_pressure_signal"] = safe_binary_column(
        df_pressure_windows,
        "positive_net_intake_flag"
    )
else:
    df_pressure_windows["intake_pressure_signal"] = (
        safe_numeric(df_pressure_windows["net_daily_intake"]).fillna(0) > 0
    ).astype(int)

# ------------------------------------------------------------
# 6. Create combined daily strain score
# ------------------------------------------------------------

strain_signal_columns = [
    "rolling_pressure_signal",
    "volatility_pressure_signal",
    "pressure_spike_signal",
    "pressure_spike_risk_signal",
    "high_load_signal",
    "backlog_pressure_signal",
    "logical_anomaly_signal",
    "intake_pressure_signal"
]

df_pressure_windows["daily_pressure_signal_count"] = df_pressure_windows[strain_signal_columns].sum(axis=1)

df_pressure_windows["daily_strain_score"] = (
    df_pressure_windows["rolling_pressure_signal"] * 2
    + df_pressure_windows["volatility_pressure_signal"] * 1
    + df_pressure_windows["pressure_spike_signal"] * 2
    + df_pressure_windows["pressure_spike_risk_signal"] * 2
    + df_pressure_windows["high_load_signal"] * 2
    + df_pressure_windows["backlog_pressure_signal"] * 2
    + df_pressure_windows["logical_anomaly_signal"] * 1
    + df_pressure_windows["intake_pressure_signal"] * 1
)

# ------------------------------------------------------------
# 7. Classify daily strain level
# ------------------------------------------------------------

def classify_daily_strain(score):
    if score >= 9:
        return "Critical Daily Strain"
    elif score >= 6:
        return "High Daily Strain"
    elif score >= 3:
        return "Moderate Daily Strain"
    else:
        return "Normal Daily Pressure"


df_pressure_windows["daily_strain_level"] = df_pressure_windows["daily_strain_score"].apply(classify_daily_strain)

df_pressure_windows["high_pressure_day_flag"] = (
    df_pressure_windows["daily_strain_score"] >= 3
).astype(int)

df_pressure_windows["severe_pressure_day_flag"] = (
    df_pressure_windows["daily_strain_score"] >= 6
).astype(int)

print("\nDaily strain indicators created successfully.")

# ------------------------------------------------------------
# 8. Detect consecutive high-pressure windows
# ------------------------------------------------------------

df_pressure_windows["pressure_window_break"] = (
    df_pressure_windows["high_pressure_day_flag"]
    .ne(df_pressure_windows["high_pressure_day_flag"].shift())
).cumsum()

df_pressure_windows["pressure_window_id"] = np.where(
    df_pressure_windows["high_pressure_day_flag"] == 1,
    df_pressure_windows["pressure_window_break"],
    np.nan
)

# Keep only high-pressure days for window aggregation
df_high_pressure_days = df_pressure_windows[
    df_pressure_windows["high_pressure_day_flag"] == 1
].copy()

minimum_window_days = 3

if df_high_pressure_days.empty:
    df_prolonged_pressure_windows = pd.DataFrame(
        columns=[
            "pressure_window_id",
            "window_start_date",
            "window_end_date",
            "window_duration_days",
            "reporting_days",
            "avg_system_load",
            "peak_system_load",
            "avg_daily_strain_score",
            "max_daily_strain_score",
            "severe_pressure_days",
            "backlog_pressure_days",
            "spike_pressure_days",
            "volatility_pressure_days",
            "total_net_intake",
            "avg_net_daily_intake",
            "window_pressure_category"
        ]
    )
else:
    df_prolonged_pressure_windows = (
        df_high_pressure_days
        .groupby("pressure_window_id", dropna=True)
        .agg(
            window_start_date=("Date", "min"),
            window_end_date=("Date", "max"),
            reporting_days=("Date", "count"),
            avg_system_load=("total_system_load", "mean"),
            peak_system_load=("total_system_load", "max"),
            avg_daily_strain_score=("daily_strain_score", "mean"),
            max_daily_strain_score=("daily_strain_score", "max"),
            severe_pressure_days=("severe_pressure_day_flag", "sum"),
            backlog_pressure_days=("backlog_pressure_signal", "sum"),
            spike_pressure_days=("pressure_spike_signal", "sum"),
            volatility_pressure_days=("volatility_pressure_signal", "sum"),
            total_net_intake=("net_daily_intake", "sum"),
            avg_net_daily_intake=("net_daily_intake", "mean")
        )
        .reset_index()
    )

    df_prolonged_pressure_windows["window_duration_days"] = (
        df_prolonged_pressure_windows["window_end_date"]
        - df_prolonged_pressure_windows["window_start_date"]
    ).dt.days + 1

    df_prolonged_pressure_windows = df_prolonged_pressure_windows[
        df_prolonged_pressure_windows["reporting_days"] >= minimum_window_days
    ].copy()

    def classify_window_pressure(row):
        if row["max_daily_strain_score"] >= 9 or row["severe_pressure_days"] >= 5:
            return "Critical Prolonged Strain Window"
        elif row["avg_daily_strain_score"] >= 6 or row["severe_pressure_days"] >= 2:
            return "High Prolonged Strain Window"
        elif row["reporting_days"] >= 7:
            return "Extended Moderate Pressure Window"
        else:
            return "Short Prolonged Pressure Window"

    if not df_prolonged_pressure_windows.empty:
        df_prolonged_pressure_windows["window_pressure_category"] = (
            df_prolonged_pressure_windows.apply(classify_window_pressure, axis=1)
        )

        numeric_window_columns = [
            "avg_system_load",
            "avg_daily_strain_score",
            "avg_net_daily_intake"
        ]

        for col in numeric_window_columns:
            df_prolonged_pressure_windows[col] = df_prolonged_pressure_windows[col].round(2)

# ------------------------------------------------------------
# 9. Add prolonged window flags back to daily dataset
# ------------------------------------------------------------

df_pressure_windows["prolonged_pressure_window_flag"] = 0
df_pressure_windows["prolonged_pressure_window_category"] = "No Prolonged Pressure Window"

if not df_prolonged_pressure_windows.empty:
    valid_window_ids = df_prolonged_pressure_windows["pressure_window_id"].tolist()

    df_pressure_windows["prolonged_pressure_window_flag"] = (
        df_pressure_windows["pressure_window_id"].isin(valid_window_ids)
    ).astype(int)

    window_category_map = dict(
        zip(
            df_prolonged_pressure_windows["pressure_window_id"],
            df_prolonged_pressure_windows["window_pressure_category"]
        )
    )

    df_pressure_windows["prolonged_pressure_window_category"] = (
        df_pressure_windows["pressure_window_id"]
        .map(window_category_map)
        .fillna("No Prolonged Pressure Window")
    )

# ------------------------------------------------------------
# 10. Create daily pressure window level distribution
# ------------------------------------------------------------

df_daily_strain_distribution = (
    df_pressure_windows["daily_strain_level"]
    .value_counts(dropna=False)
    .reset_index()
)

df_daily_strain_distribution.columns = [
    "Daily Strain Level",
    "Record Count"
]

df_daily_strain_distribution["Record Percentage"] = (
    df_daily_strain_distribution["Record Count"]
    / len(df_pressure_windows)
    * 100
).round(2)

# ------------------------------------------------------------
# 11. Create prolonged window category distribution
# ------------------------------------------------------------

df_pressure_window_category_distribution = (
    df_pressure_windows["prolonged_pressure_window_category"]
    .value_counts(dropna=False)
    .reset_index()
)

df_pressure_window_category_distribution.columns = [
    "Prolonged Pressure Window Category",
    "Record Count"
]

df_pressure_window_category_distribution["Record Percentage"] = (
    df_pressure_window_category_distribution["Record Count"]
    / len(df_pressure_windows)
    * 100
).round(2)

# ------------------------------------------------------------
# 12. Create Step 5E validation summary
# ------------------------------------------------------------

new_step_5e_columns = [
    "rolling_pressure_signal",
    "volatility_pressure_signal",
    "pressure_spike_signal",
    "pressure_spike_risk_signal",
    "high_load_signal",
    "backlog_pressure_signal",
    "logical_anomaly_signal",
    "intake_pressure_signal",
    "daily_pressure_signal_count",
    "daily_strain_score",
    "daily_strain_level",
    "high_pressure_day_flag",
    "severe_pressure_day_flag",
    "pressure_window_break",
    "pressure_window_id",
    "prolonged_pressure_window_flag",
    "prolonged_pressure_window_category"
]

missing_new_step_5e_columns = [
    col for col in new_step_5e_columns
    if col not in df_pressure_windows.columns
]

missing_values_new_step_5e = int(
    df_pressure_windows[new_step_5e_columns]
    .isna()
    .sum()
    .sum()
)

df_step_5e_validation_summary = pd.DataFrame({
    "Step 5E Check": [
        "Input Dataset Source",
        "Rows Checked",
        "Columns Before Step 5E",
        "Columns After Step 5E",
        "New Window Columns Created",
        "Missing Required Input Columns",
        "Missing New Window Columns",
        "Missing Values in New Window Columns",
        "Date Parse Errors",
        "Duplicate Dates",
        "Chronological Order",
        "High-Pressure Days",
        "Severe Pressure Days",
        "Minimum Window Days Used",
        "Prolonged Pressure Windows Detected",
        "Rows Deleted or Modified"
    ],
    "Result": [
        input_source_name,
        df_pressure_windows_base.shape[0],
        df_pressure_windows_base.shape[1],
        df_pressure_windows.shape[1],
        len(new_step_5e_columns),
        len(missing_base_columns),
        len(missing_new_step_5e_columns),
        missing_values_new_step_5e,
        date_parse_errors,
        duplicate_dates,
        chronological_order,
        int(df_pressure_windows["high_pressure_day_flag"].sum()),
        int(df_pressure_windows["severe_pressure_day_flag"].sum()),
        minimum_window_days,
        len(df_prolonged_pressure_windows),
        df_pressure_windows_base.shape[0] - df_pressure_windows.shape[0]
    ]
})

# ------------------------------------------------------------
# 13. Preview columns
# ------------------------------------------------------------

pressure_window_preview_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "daily_pressure_signal_count",
    "daily_strain_score",
    "daily_strain_level",
    "high_pressure_day_flag",
    "severe_pressure_day_flag",
    "pressure_window_id",
    "prolonged_pressure_window_flag",
    "prolonged_pressure_window_category"
]

pressure_window_preview_columns = [
    col for col in pressure_window_preview_columns
    if col in df_pressure_windows.columns
]

# ------------------------------------------------------------
# 14. Final outputs
# ------------------------------------------------------------

df_pressure_windows_daily = df_pressure_windows.copy()
df_prolonged_strain_windows = df_prolonged_pressure_windows.copy()

print("\nDaily strain level distribution:")
display(df_daily_strain_distribution)

print("\nProlonged pressure window category distribution:")
display(df_pressure_window_category_distribution)

print("\nProlonged strain / high-pressure window summary:")
display(df_prolonged_strain_windows)

print("\nStep 5E validation summary:")
display(df_step_5e_validation_summary)

print("\nPressure window daily dataset preview:")
display(df_pressure_windows_daily[pressure_window_preview_columns].head(10))

print("\nPressure window daily dataset tail:")
display(df_pressure_windows_daily[pressure_window_preview_columns].tail(10))

print("\nStep 5E completed successfully.")
print("Prolonged strain / high-pressure windows are ready for Step 5F.")
```

## Step 5F — Create Final Pressure & Stress Classification Summary

This step creates the final pressure and stress classification layer by combining rolling pressure, volatility, spike indicators, and prolonged pressure window signals.

The objective is to convert daily operational pressure patterns into an executive-ready classification system for identifying normal, early warning, moderate stress, high stress, and critical pressure days.

This step preserves all existing rows and columns from previous steps.

```python
# ============================================================
# Step 5F - Create Final Pressure & Stress Classification Summary
# ============================================================

import pandas as pd
import numpy as np

print("Step 5F - Create Final Pressure & Stress Classification Summary")
print("=" * 75)

# ------------------------------------------------------------
# 1. Select safest available input dataset
# ------------------------------------------------------------

if "df_pressure_windows_daily" in globals():
    df_pressure_final = df_pressure_windows_daily.copy()
    input_dataset_source = "df_pressure_windows_daily"
elif "df_pressure_spike" in globals():
    df_pressure_final = df_pressure_spike.copy()
    input_dataset_source = "df_pressure_spike"
elif "df_pressure_volatility" in globals():
    df_pressure_final = df_pressure_volatility.copy()
    input_dataset_source = "df_pressure_volatility"
elif "df_pressure_rolling" in globals():
    df_pressure_final = df_pressure_rolling.copy()
    input_dataset_source = "df_pressure_rolling"
else:
    raise NameError(
        "No valid Step 5 input dataset found. Please run Step 5B to Step 5E before Step 5F."
    )

print("Final pressure working copy created successfully.")
print(f"Input dataset source: {input_dataset_source}")
print(f"Input dataset shape: {df_pressure_final.shape[0]} rows × {df_pressure_final.shape[1]} columns")

rows_before_step_5f = df_pressure_final.shape[0]
columns_before_step_5f = df_pressure_final.shape[1]

# ------------------------------------------------------------
# 2. Basic validation
# ------------------------------------------------------------

if "Date" not in df_pressure_final.columns:
    raise KeyError("Required column missing: Date")

df_pressure_final["Date"] = pd.to_datetime(df_pressure_final["Date"], errors="coerce")

date_parse_errors = int(df_pressure_final["Date"].isna().sum())

df_pressure_final = df_pressure_final.sort_values("Date").reset_index(drop=True)

duplicate_dates = int(df_pressure_final["Date"].duplicated().sum())
chronological_order = bool(df_pressure_final["Date"].is_monotonic_increasing)

required_base_columns = [
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score"
]

missing_required_base_columns = [
    col for col in required_base_columns if col not in df_pressure_final.columns
]

if missing_required_base_columns:
    raise KeyError(f"Missing required base columns for Step 5F: {missing_required_base_columns}")

print(f"Missing Step 5F base input columns: {len(missing_required_base_columns)}")
print(f"Date parse errors: {date_parse_errors}")
print(f"Dataset chronological order: {chronological_order}")
print(f"Duplicate dates: {duplicate_dates}")

# ------------------------------------------------------------
# 3. Helper functions
# ------------------------------------------------------------

def get_numeric_series(df, column_name, default_value=0):
    """
    Safely returns a numeric series.
    If column is missing, returns a default numeric series.
    """
    if column_name in df.columns:
        return pd.to_numeric(df[column_name], errors="coerce").fillna(default_value)
    return pd.Series(default_value, index=df.index)


def normalize_to_10(series):
    """
    Converts any numeric signal into a 0-10 scale using min-max normalization.
    Handles flat or empty series safely.
    """
    series = pd.to_numeric(series, errors="coerce").fillna(0)

    min_value = series.min()
    max_value = series.max()

    if pd.isna(min_value) or pd.isna(max_value) or max_value == min_value:
        return pd.Series(0, index=series.index)

    return ((series - min_value) / (max_value - min_value) * 10).round(2)


def binary_score(series, score_value=10):
    """
    Converts binary/boolean flags into numeric score.
    """
    return np.where(pd.to_numeric(series, errors="coerce").fillna(0) > 0, score_value, 0)


# ------------------------------------------------------------
# 4. Build calibrated component scores
# ------------------------------------------------------------

rolling_pressure_score = get_numeric_series(df_pressure_final, "rolling_pressure_score", 0)
volatility_pressure_score = get_numeric_series(df_pressure_final, "volatility_pressure_score", 0)
pressure_spike_score = get_numeric_series(df_pressure_final, "pressure_spike_score", 0)
daily_strain_score = get_numeric_series(df_pressure_final, "daily_strain_score", 0)

high_pressure_day_flag = get_numeric_series(df_pressure_final, "high_pressure_day_flag", 0)
severe_pressure_day_flag = get_numeric_series(df_pressure_final, "severe_pressure_day_flag", 0)

prolonged_window_flag = get_numeric_series(df_pressure_final, "prolonged_pressure_window_flag", 0)
critical_window_flag = get_numeric_series(df_pressure_final, "critical_prolonged_window_flag", 0)

# Fallback window scoring from category column if direct flags are not available
if "prolonged_pressure_window_category" in df_pressure_final.columns:
    window_category_score_map = {
        "No Prolonged Pressure Window": 0,
        "Short Prolonged Pressure Window": 3,
        "High Prolonged Strain Window": 7,
        "Critical Prolonged Strain Window": 10
    }

    prolonged_category_score = (
        df_pressure_final["prolonged_pressure_window_category"]
        .map(window_category_score_map)
        .fillna(0)
    )
else:
    prolonged_category_score = pd.Series(0, index=df_pressure_final.index)

# Normalize raw operational intensity signals
system_load_intensity_score = normalize_to_10(df_pressure_final["total_system_load"])
net_intake_pressure_score = normalize_to_10(df_pressure_final["net_daily_intake"].abs())
growth_pressure_score = normalize_to_10(df_pressure_final["care_load_growth_rate_pct"].abs())
backlog_pressure_score = normalize_to_10(df_pressure_final["backlog_severity_score"])

# Normalize existing Step 5 metrics
rolling_pressure_score_norm = normalize_to_10(rolling_pressure_score)
volatility_pressure_score_norm = normalize_to_10(volatility_pressure_score)
pressure_spike_score_norm = normalize_to_10(pressure_spike_score)
daily_strain_score_norm = normalize_to_10(daily_strain_score)

# Window/flag scores
high_pressure_day_score = binary_score(high_pressure_day_flag, 6)
severe_pressure_day_score = binary_score(severe_pressure_day_flag, 10)
prolonged_window_score = np.maximum(
    binary_score(prolonged_window_flag, 7),
    prolonged_category_score
)
critical_window_score = binary_score(critical_window_flag, 10)

# ------------------------------------------------------------
# 5. Final recalibrated pressure stress score
# ------------------------------------------------------------

df_pressure_final["rolling_pressure_component_score"] = rolling_pressure_score_norm
df_pressure_final["volatility_pressure_component_score"] = volatility_pressure_score_norm
df_pressure_final["pressure_spike_component_score"] = pressure_spike_score_norm
df_pressure_final["daily_strain_component_score"] = daily_strain_score_norm
df_pressure_final["load_intensity_component_score"] = system_load_intensity_score
df_pressure_final["net_intake_component_score"] = net_intake_pressure_score
df_pressure_final["growth_pressure_component_score"] = growth_pressure_score
df_pressure_final["backlog_component_score"] = backlog_pressure_score
df_pressure_final["window_pressure_component_score"] = prolonged_window_score
df_pressure_final["critical_window_component_score"] = critical_window_score

df_pressure_final["final_pressure_stress_score"] = (
    df_pressure_final["daily_strain_component_score"] * 0.22
    + df_pressure_final["volatility_pressure_component_score"] * 0.18
    + df_pressure_final["pressure_spike_component_score"] * 0.16
    + df_pressure_final["rolling_pressure_component_score"] * 0.14
    + df_pressure_final["load_intensity_component_score"] * 0.10
    + df_pressure_final["net_intake_component_score"] * 0.06
    + df_pressure_final["growth_pressure_component_score"] * 0.05
    + df_pressure_final["backlog_component_score"] * 0.05
    + df_pressure_final["window_pressure_component_score"] * 0.03
    + df_pressure_final["critical_window_component_score"] * 0.01
).round(2)

# ------------------------------------------------------------
# 6. Dynamic percentile-based recalibration thresholds
# ------------------------------------------------------------

score_series = df_pressure_final["final_pressure_stress_score"].copy()

non_zero_scores = score_series[score_series > 0]

if len(non_zero_scores) >= 10:
    q40 = non_zero_scores.quantile(0.40)
    q65 = non_zero_scores.quantile(0.65)
    q82 = non_zero_scores.quantile(0.82)
    q94 = non_zero_scores.quantile(0.94)
else:
    q40 = score_series.quantile(0.40)
    q65 = score_series.quantile(0.65)
    q82 = score_series.quantile(0.82)
    q94 = score_series.quantile(0.94)

# Make sure thresholds are stable and increasing
threshold_early = max(float(q40), 0.75)
threshold_moderate = max(float(q65), threshold_early + 0.25)
threshold_high = max(float(q82), threshold_moderate + 0.25)
threshold_critical = max(float(q94), threshold_high + 0.25)

print("\nFinal pressure stress recalibration thresholds:")
print(f"Early pressure threshold: {threshold_early:.2f}")
print(f"Moderate stress threshold: {threshold_moderate:.2f}")
print(f"High stress threshold: {threshold_high:.2f}")
print(f"Critical pressure threshold: {threshold_critical:.2f}")

# ------------------------------------------------------------
# 7. Final pressure stress classification
# ------------------------------------------------------------

def classify_final_pressure(row):
    score = row["final_pressure_stress_score"]

    severe_flag = row.get("severe_pressure_day_flag", 0)
    high_flag = row.get("high_pressure_day_flag", 0)
    spike_score = row.get("pressure_spike_score", 0)
    strain_score = row.get("daily_strain_score", 0)
    backlog_score = row.get("backlog_severity_score", 0)

    critical_window = row.get("critical_prolonged_window_flag", 0)
    prolonged_window = row.get("prolonged_pressure_window_flag", 0)

    if (
        score >= threshold_critical
        or critical_window == 1
        or (severe_flag == 1 and spike_score >= 6)
        or (strain_score >= 8 and backlog_score >= 2)
    ):
        return "Critical System Pressure"

    elif (
        score >= threshold_high
        or severe_flag == 1
        or (high_flag == 1 and spike_score >= 4)
        or (strain_score >= 6 and prolonged_window == 1)
    ):
        return "High System Stress"

    elif (
        score >= threshold_moderate
        or high_flag == 1
        or spike_score >= 4
        or strain_score >= 4
        or backlog_score >= 2
    ):
        return "Moderate System Stress"

    elif (
        score >= threshold_early
        or spike_score > 0
        or strain_score > 0
        or backlog_score > 0
    ):
        return "Early Pressure Watch"

    else:
        return "Normal Operating Pressure"


df_pressure_final["final_pressure_stress_level"] = df_pressure_final.apply(
    classify_final_pressure,
    axis=1
)

# ------------------------------------------------------------
# 8. Operational response category
# ------------------------------------------------------------

operational_response_map = {
    "Normal Operating Pressure": "Routine Monitoring",
    "Early Pressure Watch": "Early Warning Review",
    "Moderate System Stress": "Operational Watch Required",
    "High System Stress": "Priority Monitoring Required",
    "Critical System Pressure": "Immediate Escalation Required"
}

df_pressure_final["operational_response_category"] = (
    df_pressure_final["final_pressure_stress_level"]
    .map(operational_response_map)
    .fillna("Review Required")
)

# ------------------------------------------------------------
# 9. Primary pressure driver classification
# ------------------------------------------------------------

def classify_primary_pressure_driver(row):
    driver_scores = {
        "Rolling Load Pressure": row.get("rolling_pressure_component_score", 0),
        "Volatility Pressure": row.get("volatility_pressure_component_score", 0),
        "Sudden Spike Pressure": row.get("pressure_spike_component_score", 0),
        "Daily Strain Pressure": row.get("daily_strain_component_score", 0),
        "Backlog Severity Pressure": row.get("backlog_component_score", 0),
        "Prolonged Window Pressure": row.get("window_pressure_component_score", 0)
    }

    max_driver = max(driver_scores, key=driver_scores.get)
    max_score = driver_scores[max_driver]

    if max_score <= 0:
        return "No Major Pressure Driver"

    return max_driver


df_pressure_final["primary_pressure_driver"] = df_pressure_final.apply(
    classify_primary_pressure_driver,
    axis=1
)

# ------------------------------------------------------------
# 10. Final pressure priority rank
# ------------------------------------------------------------

priority_rank_map = {
    "Normal Operating Pressure": 1,
    "Early Pressure Watch": 2,
    "Moderate System Stress": 3,
    "High System Stress": 4,
    "Critical System Pressure": 5
}

df_pressure_final["final_pressure_priority_rank"] = (
    df_pressure_final["final_pressure_stress_level"]
    .map(priority_rank_map)
    .fillna(0)
    .astype(int)
)

# ------------------------------------------------------------
# 11. Summary tables
# ------------------------------------------------------------

final_pressure_stress_distribution = (
    df_pressure_final["final_pressure_stress_level"]
    .value_counts()
    .reindex(
        [
            "Normal Operating Pressure",
            "Early Pressure Watch",
            "Moderate System Stress",
            "High System Stress",
            "Critical System Pressure"
        ],
        fill_value=0
    )
    .reset_index()
)

final_pressure_stress_distribution.columns = [
    "Final Pressure Stress Level",
    "Record Count"
]

final_pressure_stress_distribution["Record Percentage"] = (
    final_pressure_stress_distribution["Record Count"]
    / len(df_pressure_final)
    * 100
).round(2)

operational_response_distribution = (
    df_pressure_final["operational_response_category"]
    .value_counts()
    .reindex(
        [
            "Routine Monitoring",
            "Early Warning Review",
            "Operational Watch Required",
            "Priority Monitoring Required",
            "Immediate Escalation Required"
        ],
        fill_value=0
    )
    .reset_index()
)

operational_response_distribution.columns = [
    "Operational Response Category",
    "Record Count"
]

operational_response_distribution["Record Percentage"] = (
    operational_response_distribution["Record Count"]
    / len(df_pressure_final)
    * 100
).round(2)

primary_pressure_driver_summary = (
    df_pressure_final["primary_pressure_driver"]
    .value_counts()
    .reset_index()
)

primary_pressure_driver_summary.columns = [
    "Primary Pressure Driver",
    "Record Count"
]

primary_pressure_driver_summary["Record Percentage"] = (
    primary_pressure_driver_summary["Record Count"]
    / len(df_pressure_final)
    * 100
).round(2)

executive_final_pressure_summary = pd.DataFrame({
    "Final Pressure Summary Metric": [
        "Input Dataset Source",
        "Total Reporting Days",
        "Normal Operating Pressure Days",
        "Early Pressure Watch Days",
        "Moderate System Stress Days",
        "High System Stress Days",
        "Critical System Pressure Days",
        "Average Final Pressure Stress Score",
        "Maximum Final Pressure Stress Score",
        "Highest Pressure Date",
        "Rows Deleted or Modified"
    ],
    "Result": [
        input_dataset_source,
        len(df_pressure_final),
        int((df_pressure_final["final_pressure_stress_level"] == "Normal Operating Pressure").sum()),
        int((df_pressure_final["final_pressure_stress_level"] == "Early Pressure Watch").sum()),
        int((df_pressure_final["final_pressure_stress_level"] == "Moderate System Stress").sum()),
        int((df_pressure_final["final_pressure_stress_level"] == "High System Stress").sum()),
        int((df_pressure_final["final_pressure_stress_level"] == "Critical System Pressure").sum()),
        round(df_pressure_final["final_pressure_stress_score"].mean(), 2),
        round(df_pressure_final["final_pressure_stress_score"].max(), 2),
        df_pressure_final.loc[
            df_pressure_final["final_pressure_stress_score"].idxmax(),
            "Date"
        ].date(),
        rows_before_step_5f - df_pressure_final.shape[0]
    ]
})

# ------------------------------------------------------------
# 12. Validation summary
# ------------------------------------------------------------

new_step_5f_columns = [
    "rolling_pressure_component_score",
    "volatility_pressure_component_score",
    "pressure_spike_component_score",
    "daily_strain_component_score",
    "load_intensity_component_score",
    "net_intake_component_score",
    "growth_pressure_component_score",
    "backlog_component_score",
    "window_pressure_component_score",
    "critical_window_component_score",
    "final_pressure_stress_score",
    "final_pressure_stress_level",
    "operational_response_category",
    "primary_pressure_driver",
    "final_pressure_priority_rank"
]

missing_new_step_5f_columns = [
    col for col in new_step_5f_columns if col not in df_pressure_final.columns
]

missing_values_new_step_5f_columns = int(
    df_pressure_final[new_step_5f_columns].isna().sum().sum()
)

step_5f_validation_summary = pd.DataFrame({
    "Step 5F Check": [
        "Input Dataset Source",
        "Rows Checked",
        "Columns Before Step 5F",
        "Columns After Step 5F",
        "New Final Classification Columns Created",
        "Missing Required Base Columns",
        "Missing New Final Columns",
        "Missing Values in New Final Columns",
        "Date Parse Errors",
        "Duplicate Dates",
        "Chronological Order",
        "Rows Deleted or Modified"
    ],
    "Result": [
        input_dataset_source,
        df_pressure_final.shape[0],
        columns_before_step_5f,
        df_pressure_final.shape[1],
        len(new_step_5f_columns),
        len(missing_required_base_columns),
        len(missing_new_step_5f_columns),
        missing_values_new_step_5f_columns,
        date_parse_errors,
        duplicate_dates,
        chronological_order,
        rows_before_step_5f - df_pressure_final.shape[0]
    ]
})

# ------------------------------------------------------------
# 13. Display summaries
# ------------------------------------------------------------

print("\nFinal pressure classification created successfully.")

print("\nFinal pressure stress level distribution:")
display(final_pressure_stress_distribution)

print("\nOperational response category distribution:")
display(operational_response_distribution)

print("\nPrimary pressure driver summary:")
display(primary_pressure_driver_summary)

print("\nExecutive final pressure classification summary:")
display(executive_final_pressure_summary)

print("\nStep 5F validation summary:")
display(step_5f_validation_summary)

# ------------------------------------------------------------
# 14. Display final dataset preview
# ------------------------------------------------------------

final_pressure_preview_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "rolling_pressure_score",
    "volatility_pressure_score",
    "pressure_spike_score",
    "daily_strain_score",
    "final_pressure_stress_score",
    "final_pressure_stress_level",
    "operational_response_category",
    "primary_pressure_driver",
    "final_pressure_priority_rank"
]

final_pressure_preview_columns = [
    col for col in final_pressure_preview_columns if col in df_pressure_final.columns
]

print("\nFinal pressure classification dataset preview:")
display(df_pressure_final[final_pressure_preview_columns].head(10))

print("\nFinal pressure classification dataset tail:")
display(df_pressure_final[final_pressure_preview_columns].tail(10))

print("\nStep 5F completed successfully.")
print("Final pressure and stress classification summary is ready for Step 5G.")
```

## Step 5G — Export Final Pressure & Stress Classification Dataset

In this step, the final pressure and stress classification dataset is exported for downstream analysis, visualization, and dashboard preparation.

This step uses the final classified dataset created in Step 5F and saves it as a structured CSV file inside the processed data directory.

### Objectives

- Export the final pressure and stress classification dataset.
- Preserve all engineered rolling, volatility, spike, prolonged strain, and final classification metrics.
- Create a compact export validation summary.
- Confirm that the exported dataset has the expected row count, column count, and no duplicate dates.
- Prepare the dataset for the next notebook/dashboard stage.

### Output

The final exported file will be saved as:

`data/processed/final_pressure_stress_classification_dataset.csv`

```python
# ============================================================
# Step 5G — Export Final Pressure & Stress Classification Dataset
# ============================================================

print("Step 5G - Export Final Pressure & Stress Classification Dataset")
print("=" * 75)

# ------------------------------------------------------------
# 1. Import required libraries
# ------------------------------------------------------------

import os
import pandas as pd


# ------------------------------------------------------------
# 2. Validate Step 5F final dataset availability
# ------------------------------------------------------------

required_step_5g_input = "df_pressure_final"

if required_step_5g_input not in globals():
    raise NameError(
        "Step 5G Error: df_pressure_final is not available. "
        "Please run Step 5F before Step 5G."
    )

if not isinstance(df_pressure_final, pd.DataFrame):
    raise TypeError(
        "Step 5G Error: df_pressure_final exists but is not a pandas DataFrame."
    )

if df_pressure_final.empty:
    raise ValueError(
        "Step 5G Error: df_pressure_final is empty. "
        "Please verify Step 5F output before exporting."
    )

print("Final pressure dataset found successfully.")
print(f"Input dataset shape: {df_pressure_final.shape[0]} rows × {df_pressure_final.shape[1]} columns")


# ------------------------------------------------------------
# 3. Create safe export copy
# ------------------------------------------------------------

df_pressure_export = df_pressure_final.copy()

print("Export working copy created successfully.")


# ------------------------------------------------------------
# 4. Ensure Date column is clean and chronological
# ------------------------------------------------------------

if "Date" not in df_pressure_export.columns:
    raise KeyError(
        "Step 5G Error: Date column is missing from df_pressure_final."
    )

df_pressure_export["Date"] = pd.to_datetime(df_pressure_export["Date"], errors="coerce")

date_parse_errors = int(df_pressure_export["Date"].isna().sum())

if date_parse_errors > 0:
    raise ValueError(
        f"Step 5G Error: {date_parse_errors} Date values could not be parsed."
    )

df_pressure_export = df_pressure_export.sort_values("Date").reset_index(drop=True)

duplicate_dates = int(df_pressure_export["Date"].duplicated().sum())
chronological_order_status = bool(df_pressure_export["Date"].is_monotonic_increasing)

print(f"Date parse errors: {date_parse_errors}")
print(f"Duplicate dates: {duplicate_dates}")
print(f"Dataset chronological order: {chronological_order_status}")


# ------------------------------------------------------------
# 5. Validate important final classification columns
# ------------------------------------------------------------

required_final_export_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "final_pressure_stress_score",
    "final_pressure_stress_level",
    "operational_response_category",
    "primary_pressure_driver"
]

missing_final_export_columns = [
    col for col in required_final_export_columns
    if col not in df_pressure_export.columns
]

if missing_final_export_columns:
    raise KeyError(
        "Step 5G Error: Missing required final export columns: "
        f"{missing_final_export_columns}"
    )

missing_values_in_required_columns = int(
    df_pressure_export[required_final_export_columns].isna().sum().sum()
)

print(f"Missing required final export columns: {len(missing_final_export_columns)}")
print(f"Missing values in required export columns: {missing_values_in_required_columns}")


# ------------------------------------------------------------
# 6. Create export directory
# ------------------------------------------------------------

processed_output_dir = os.path.join("data", "processed")
os.makedirs(processed_output_dir, exist_ok=True)

print(f"Processed output directory ready: {processed_output_dir}")


# ------------------------------------------------------------
# 7. Define export file paths
# ------------------------------------------------------------

final_pressure_export_path = os.path.join(
    processed_output_dir,
    "final_pressure_stress_classification_dataset.csv"
)

final_pressure_summary_export_path = os.path.join(
    processed_output_dir,
    "final_pressure_stress_export_summary.csv"
)

final_pressure_level_distribution_path = os.path.join(
    processed_output_dir,
    "final_pressure_stress_level_distribution.csv"
)

operational_response_distribution_path = os.path.join(
    processed_output_dir,
    "operational_response_category_distribution.csv"
)

primary_pressure_driver_distribution_path = os.path.join(
    processed_output_dir,
    "primary_pressure_driver_distribution.csv"
)


# ------------------------------------------------------------
# 8. Export final full dataset
# ------------------------------------------------------------

df_pressure_export.to_csv(final_pressure_export_path, index=False)

print("Final pressure and stress classification dataset exported successfully.")
print(f"Export path: {final_pressure_export_path}")


# ------------------------------------------------------------
# 9. Create distribution summary exports
# ------------------------------------------------------------

final_pressure_level_distribution = (
    df_pressure_export["final_pressure_stress_level"]
    .value_counts(dropna=False)
    .reset_index()
)

final_pressure_level_distribution.columns = [
    "Final Pressure Stress Level",
    "Record Count"
]

final_pressure_level_distribution["Record Percentage"] = (
    final_pressure_level_distribution["Record Count"] / len(df_pressure_export) * 100
).round(2)

final_pressure_level_distribution.to_csv(
    final_pressure_level_distribution_path,
    index=False
)


operational_response_distribution = (
    df_pressure_export["operational_response_category"]
    .value_counts(dropna=False)
    .reset_index()
)

operational_response_distribution.columns = [
    "Operational Response Category",
    "Record Count"
]

operational_response_distribution["Record Percentage"] = (
    operational_response_distribution["Record Count"] / len(df_pressure_export) * 100
).round(2)

operational_response_distribution.to_csv(
    operational_response_distribution_path,
    index=False
)


primary_pressure_driver_distribution = (
    df_pressure_export["primary_pressure_driver"]
    .value_counts(dropna=False)
    .reset_index()
)

primary_pressure_driver_distribution.columns = [
    "Primary Pressure Driver",
    "Record Count"
]

primary_pressure_driver_distribution["Record Percentage"] = (
    primary_pressure_driver_distribution["Record Count"] / len(df_pressure_export) * 100
).round(2)

primary_pressure_driver_distribution.to_csv(
    primary_pressure_driver_distribution_path,
    index=False
)

print("Final distribution summary files exported successfully.")


# ------------------------------------------------------------
# 10. Create export validation summary
# ------------------------------------------------------------

final_export_summary = pd.DataFrame({
    "Step 5G Export Check": [
        "Input Dataset Source",
        "Exported Dataset Rows",
        "Exported Dataset Columns",
        "Date Parse Errors",
        "Duplicate Dates",
        "Chronological Order",
        "Missing Required Export Columns",
        "Missing Values in Required Export Columns",
        "Final Pressure Stress Levels",
        "Operational Response Categories",
        "Primary Pressure Drivers",
        "Full Dataset Export Path",
        "Stress Level Distribution Export Path",
        "Operational Response Distribution Export Path",
        "Primary Pressure Driver Distribution Export Path"
    ],
    "Result": [
        "df_pressure_final",
        df_pressure_export.shape[0],
        df_pressure_export.shape[1],
        date_parse_errors,
        duplicate_dates,
        chronological_order_status,
        len(missing_final_export_columns),
        missing_values_in_required_columns,
        df_pressure_export["final_pressure_stress_level"].nunique(dropna=False),
        df_pressure_export["operational_response_category"].nunique(dropna=False),
        df_pressure_export["primary_pressure_driver"].nunique(dropna=False),
        final_pressure_export_path,
        final_pressure_level_distribution_path,
        operational_response_distribution_path,
        primary_pressure_driver_distribution_path
    ]
})

final_export_summary.to_csv(final_pressure_summary_export_path, index=False)

print("Step 5G export validation summary created successfully.")


# ------------------------------------------------------------
# 11. Display export validation summary
# ------------------------------------------------------------

print("\nStep 5G validation summary:")
display(final_export_summary)


# ------------------------------------------------------------
# 12. Display final distribution summaries
# ------------------------------------------------------------

print("\nFinal pressure stress level distribution:")
display(final_pressure_level_distribution)

print("\nOperational response category distribution:")
display(operational_response_distribution)

print("\nPrimary pressure driver distribution:")
display(primary_pressure_driver_distribution)


# ------------------------------------------------------------
# 13. Display final exported dataset preview
# ------------------------------------------------------------

final_export_preview_columns = [
    col for col in [
        "Date",
        "total_system_load",
        "net_daily_intake",
        "care_load_growth_rate_pct",
        "backlog_severity_score",
        "volatility_pressure_score",
        "pressure_spike_score",
        "daily_strain_score",
        "final_pressure_stress_score",
        "final_pressure_stress_level",
        "operational_response_category",
        "primary_pressure_driver"
    ]
    if col in df_pressure_export.columns
]

print("\nFinal exported dataset preview:")
display(df_pressure_export[final_export_preview_columns].head(10))

print("\nFinal exported dataset tail:")
display(df_pressure_export[final_export_preview_columns].tail(10))


# ------------------------------------------------------------
# 14. Completion message
# ------------------------------------------------------------

print("\nStep 5G completed successfully.")
print("Final pressure and stress classification dataset is exported and ready for Step 5H or the next notebook stage.")
```

```python
# ============================================================
# Step 5H — Final Notebook 05 Completion Summary / Validation Wrap-up
# ============================================================

print("Step 5H - Final Notebook 05 Completion Summary / Validation Wrap-up")
print("=" * 80)

import pandas as pd
from pathlib import Path

# ------------------------------------------------------------
# 1. Validate required final dataset from Step 5G
# ------------------------------------------------------------

required_final_objects = ["df_pressure_export"]

missing_final_objects = [
    obj for obj in required_final_objects
    if obj not in globals()
]

if missing_final_objects:
    raise NameError(
        f"Missing required final object(s) from previous step: {missing_final_objects}. "
        "Please run Step 5G before Step 5H."
    )

df_notebook_05_final = df_pressure_export.copy()

print("Final Notebook 05 dataset loaded successfully.")
print(f"Final dataset shape: {df_notebook_05_final.shape[0]} rows × {df_notebook_05_final.shape[1]} columns")

# ------------------------------------------------------------
# 2. Confirm final export path
# ------------------------------------------------------------

final_export_path = Path("data/processed/final_pressure_stress_classification_dataset.csv")

export_file_exists = final_export_path.exists()

print(f"Expected final export path: {final_export_path}")
print(f"Final exported file exists: {export_file_exists}")

if not export_file_exists:
    raise FileNotFoundError(
        "Final exported dataset file was not found. "
        "Please rerun Step 5G before completing Step 5H."
    )

# ------------------------------------------------------------
# 3. Validate key final classification columns
# ------------------------------------------------------------

expected_final_columns = [
    "Date",
    "total_system_load",
    "net_daily_intake",
    "care_load_growth_rate_pct",
    "backlog_severity_score",
    "volatility_pressure_score",
    "pressure_spike_score",
    "daily_strain_score",
    "final_pressure_stress_score",
    "final_pressure_stress_level",
    "operational_response_category",
    "primary_pressure_driver"
]

available_expected_columns = [
    col for col in expected_final_columns
    if col in df_notebook_05_final.columns
]

missing_expected_columns = [
    col for col in expected_final_columns
    if col not in df_notebook_05_final.columns
]

print(f"Expected final columns checked: {len(expected_final_columns)}")
print(f"Expected final columns available: {len(available_expected_columns)}")
print(f"Missing expected final columns: {len(missing_expected_columns)}")

if missing_expected_columns:
    print("Missing expected final columns list:")
    print(missing_expected_columns)

# ------------------------------------------------------------
# 4. Create step completion tracker
# ------------------------------------------------------------

notebook_05_step_completion_summary = pd.DataFrame({
    "Notebook Step": [
        "Step 5A",
        "Step 5B",
        "Step 5C",
        "Step 5D",
        "Step 5E",
        "Step 5F",
        "Step 5G",
        "Step 5H"
    ],
    "Step Description": [
        "Create Base Pressure and Stress Identification Dataset",
        "Create Rolling Average Pressure Metrics",
        "Create Variability / Volatility Pressure Metrics",
        "Create Pressure Spike / Sudden Change Indicators",
        "Detect Prolonged Strain / High-Pressure Windows",
        "Create Final Pressure & Stress Classification Summary",
        "Export Final Pressure & Stress Classification Dataset",
        "Final Notebook 05 Completion Summary / Validation Wrap-up"
    ],
    "Status": [
        "Completed",
        "Completed",
        "Completed",
        "Completed",
        "Completed",
        "Completed",
        "Completed",
        "Completed"
    ]
})

print("\nNotebook 05 step completion tracker:")
display(notebook_05_step_completion_summary)

# ------------------------------------------------------------
# 5. Final pressure stress distribution summary
# ------------------------------------------------------------

if "final_pressure_stress_level" in df_notebook_05_final.columns:
    final_pressure_distribution = (
        df_notebook_05_final["final_pressure_stress_level"]
        .value_counts(dropna=False)
        .reset_index()
    )

    final_pressure_distribution.columns = [
        "Final Pressure Stress Level",
        "Record Count"
    ]

    final_pressure_distribution["Record Percentage"] = (
        final_pressure_distribution["Record Count"] / len(df_notebook_05_final) * 100
    ).round(2)

    print("\nFinal pressure stress level distribution:")
    display(final_pressure_distribution)
else:
    final_pressure_distribution = pd.DataFrame()
    print("\nFinal pressure stress level distribution could not be created because column is missing.")

# ------------------------------------------------------------
# 6. Operational response distribution summary
# ------------------------------------------------------------

if "operational_response_category" in df_notebook_05_final.columns:
    operational_response_distribution = (
        df_notebook_05_final["operational_response_category"]
        .value_counts(dropna=False)
        .reset_index()
    )

    operational_response_distribution.columns = [
        "Operational Response Category",
        "Record Count"
    ]

    operational_response_distribution["Record Percentage"] = (
        operational_response_distribution["Record Count"] / len(df_notebook_05_final) * 100
    ).round(2)

    print("\nOperational response category distribution:")
    display(operational_response_distribution)
else:
    operational_response_distribution = pd.DataFrame()
    print("\nOperational response category distribution could not be created because column is missing.")

# ------------------------------------------------------------
# 7. Primary pressure driver summary
# ------------------------------------------------------------

if "primary_pressure_driver" in df_notebook_05_final.columns:
    primary_pressure_driver_summary = (
        df_notebook_05_final["primary_pressure_driver"]
        .value_counts(dropna=False)
        .reset_index()
    )

    primary_pressure_driver_summary.columns = [
        "Primary Pressure Driver",
        "Record Count"
    ]

    primary_pressure_driver_summary["Record Percentage"] = (
        primary_pressure_driver_summary["Record Count"] / len(df_notebook_05_final) * 100
    ).round(2)

    print("\nPrimary pressure driver summary:")
    display(primary_pressure_driver_summary)
else:
    primary_pressure_driver_summary = pd.DataFrame()
    print("\nPrimary pressure driver summary could not be created because column is missing.")

# ------------------------------------------------------------
# 8. Executive notebook completion summary
# ------------------------------------------------------------

total_reporting_days = len(df_notebook_05_final)

normal_days = (
    int((df_notebook_05_final["final_pressure_stress_level"] == "Normal Operating Pressure").sum())
    if "final_pressure_stress_level" in df_notebook_05_final.columns else 0
)

early_watch_days = (
    int((df_notebook_05_final["final_pressure_stress_level"] == "Early Pressure Watch").sum())
    if "final_pressure_stress_level" in df_notebook_05_final.columns else 0
)

moderate_stress_days = (
    int((df_notebook_05_final["final_pressure_stress_level"] == "Moderate System Stress").sum())
    if "final_pressure_stress_level" in df_notebook_05_final.columns else 0
)

high_stress_days = (
    int((df_notebook_05_final["final_pressure_stress_level"] == "High System Stress").sum())
    if "final_pressure_stress_level" in df_notebook_05_final.columns else 0
)

critical_pressure_days = (
    int((df_notebook_05_final["final_pressure_stress_level"] == "Critical System Pressure").sum())
    if "final_pressure_stress_level" in df_notebook_05_final.columns else 0
)

average_pressure_score = (
    round(df_notebook_05_final["final_pressure_stress_score"].mean(), 2)
    if "final_pressure_stress_score" in df_notebook_05_final.columns else None
)

maximum_pressure_score = (
    round(df_notebook_05_final["final_pressure_stress_score"].max(), 2)
    if "final_pressure_stress_score" in df_notebook_05_final.columns else None
)

highest_pressure_date = (
    df_notebook_05_final.loc[
        df_notebook_05_final["final_pressure_stress_score"].idxmax(),
        "Date"
    ]
    if "final_pressure_stress_score" in df_notebook_05_final.columns and "Date" in df_notebook_05_final.columns
    else None
)

notebook_05_final_summary = pd.DataFrame({
    "Notebook 05 Final Summary Metric": [
        "Final Dataset Name",
        "Final Export Path",
        "Total Reporting Days",
        "Total Final Columns",
        "Normal Operating Pressure Days",
        "Early Pressure Watch Days",
        "Moderate System Stress Days",
        "High System Stress Days",
        "Critical System Pressure Days",
        "Average Final Pressure Stress Score",
        "Maximum Final Pressure Stress Score",
        "Highest Pressure Date",
        "Rows Deleted or Modified",
        "Final Export File Exists",
        "Notebook 05 Completion Status",
        "Ready for Next Notebook Stage"
    ],
    "Result": [
        "df_notebook_05_final",
        str(final_export_path),
        total_reporting_days,
        df_notebook_05_final.shape[1],
        normal_days,
        early_watch_days,
        moderate_stress_days,
        high_stress_days,
        critical_pressure_days,
        average_pressure_score,
        maximum_pressure_score,
        highest_pressure_date,
        0,
        export_file_exists,
        "Completed Successfully",
        "Yes"
    ]
})

print("\nExecutive Notebook 05 final completion summary:")
display(notebook_05_final_summary)

# ------------------------------------------------------------
# 9. Final dataset preview
# ------------------------------------------------------------

final_preview_columns = [
    col for col in [
        "Date",
        "total_system_load",
        "net_daily_intake",
        "care_load_growth_rate_pct",
        "backlog_severity_score",
        "volatility_pressure_score",
        "pressure_spike_score",
        "daily_strain_score",
        "final_pressure_stress_score",
        "final_pressure_stress_level",
        "operational_response_category",
        "primary_pressure_driver"
    ]
    if col in df_notebook_05_final.columns
]

print("\nFinal Notebook 05 dataset preview:")
display(df_notebook_05_final[final_preview_columns].head(10))

print("\nFinal Notebook 05 dataset tail:")
display(df_notebook_05_final[final_preview_columns].tail(10))

# ------------------------------------------------------------
# 10. Completion message
# ------------------------------------------------------------

print("\nStep 5H completed successfully.")
print("Notebook 05 pressure and stress identification workflow is complete.")
print("Final pressure and stress classification dataset is exported and ready for the next notebook stage.")
```
